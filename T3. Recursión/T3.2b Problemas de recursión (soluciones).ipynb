## 1. Potencia n-ésima

Dado un número entero $a$ y $n \geq 0$, calcular la potencia $a^n$ mediante una función recursiva.

#### 1. Análisis del problema y declaración del perfil de la función

```python
def potencia(a, n):
    """
    Calcula la potencia n-ésima de un entero a.

    Parámetros:
        a (int): Base de la potencia.
        n (int): Exponente, n >= 0.

    Retorna:
        int: a elevado a la potencia n.
    """
```

#### 2. Análisis de casos

* **Caso base**: Si $n = 0$, matemáticamente sabemos que cualquier entidad $a$ elevada a cero resulta en $1$.
* **Caso general**: Si $n > 0$, entonces la potencia puede descomponerse en una multiplicación simple: $a^n = a^{n-1} \cdot a$.

#### 3. Transcripción

```python
if n == 0:
    return 1
else:
    return potencia(a, n - 1) * a
```

#### 4. Validación

- En la llamada recursiva `potencia(a, n - 1)`, reducimos el exponente `n` en una unidad en cada paso, acercándonos a 0, el caso base, asegurando la terminación.
- Como la precondición inicial es que $n \ge 0$, y el caso base se da cuando $n = 0$, $n$ nunca será negativo en ninguna llamada.

#### Tipo de recursión

**Recursión lineal no final**: el resultado de calcular la potencia de orden inferior $n-1$ debemos multiplicarlo por $a$.

#### Implementación completa + ejecución:

````python
def potencia(a, n):
    """
    Calcula la potencia n-ésima de un entero a.
    
    Parámetros:
        a (int): Base de la potencia.
        n (int): Exponente, n >= 0.
        
    Retorna:
        int: a elevado a la potencia n.
    """
    if n == 0:
        return 1
    else:
        return potencia(a, n - 1) * a

if __name__ == "__main__":
    base, exponente = 2, 5
    print(f"{base} elevado a {exponente} es: {potencia(base, exponente)}")
````

## 2. Resto de la división entera

Dados dos números naturales $a \geq 0$ y $b > 0$, calcular el resto de su división entera $a / b$ mediante una función recursiva.

#### 1. Análisis del problema y declaración del perfil de la función

```python
def resto(a, b):
    """
    Calcula el resto de la división entera de a entre b.

    Parámetros:
        a (int): Dividendo, a >= 0.
        b (int): Divisor, b > 0.

    Retorna:
        int: Resto de la división entera.
"""
```

#### 2. Análisis de casos

* **Caso base**: Si $a < b$, entonces el resto de la división de $a$ entre $b$ es simplemente $a$.
* **Caso general**: En otro caso ($a \ge b$), podemos restar el divisor $b$ al dividendo $a$, y el resto de la división original será igual al resto de dividir $(a - b)$ entre $b$.

#### 3. Transcripción

```python
if a < b:
    return a
else:
    return resto(a - b, b)
```

#### 4. Validación

- En la llamada recursiva `resto(a - b, b)`, el primer argumento decrece en el valor positivo $b$. Puesto que el parámetro `a` original es $\ge 0$, sucesivas sustracciones garantizan que eventualmente se alcance la condición $a < b$, alcanzando el **caso base** y finalizando la recursión.
- Dado que $b > 0$ y $a \ge 0$ inicialmente, en cada paso restamos $b$, pero siempre paramos en caso de que $(a - b) < b$, es decir, $a$ es siempre mayor o igual a cero. Por su lado, $b$ permanece inalterado en todas las llamadas; por tanto, si la condición $b > 0$ se cumple en la primera llamada, se cumplirá en todas.

#### Tipo de recursión

**Recursión lineal final**. El resultado de la llamada actual devuelve exactamente el mismo valor que la llamada recursiva anterior.

#### Implementación completa + ejecución:

````python
def resto(a, b):
    """
    Calcula el resto de la división entera de a entre b.
    
    Parámetros:
        a (int): Dividendo, a >= 0.
        b (int): Divisor, b > 0.
        
    Retorna:
        int: Resto de la división entera.
    """
    if a < b:
        return a
    else:
        return resto(a - b, b)

if __name__ == "__main__":
    a, b = 17, 5
    print(f"Resto de {a} / {b}: {resto(a, b)}")
````

## 3. Sucesión de Fibonacci

Calcular el término $n$-ésimo de la sucesión de Fibonacci (asumiendo que el término 0-ésimo toma el valor $0$, y el 1-ésimo toma el valor $1$) mediante una función recursiva.

#### 1. Análisis del problema y declaración del perfil de la función

```python
def fibonacci(n):
    """
    Calcula el término n-ésimo de la sucesión de Fibonacci.

    Parámetros:
        n (int): Posición en la sucesión, n >= 0.

    Retorna:
        int: Valor del término de Fibonacci en la posición n.
    """
```

#### 2. Análisis de casos

* **Casos base**: Si $n = 0$, el término es 0. Si $n = 1$, el término es 1. Dicho de otro modo, si $n \le 1$, la solución coincide con $n$.
* **Caso general**: Para $n > 1$, la posición en Fibonacci es igual a la suma de las dos posiciones contiguas precedentes de la sucesión: $(n-1)$-ésimo y $(n-2)$-ésimo.

#### 3. Transcripción

```python
if n <= 1:
    return n
else:
    return fibonacci(n - 1) + fibonacci(n - 2)
```

#### 4. Validación

- Dado que las llamadas recursivas decrecen $n$ en $1$ y $2$ unidades, la solución se aproxima poco a poco a los casos base. Cuando $n = 2$, el problema se resuelve combinando los dos casos base, por lo que se garantiza la terminación, y a demás, que $n$ nunca será negativo.

#### Tipo de recursión 

**Recursión múltiple**: la solución al problema depende de la realización de dos llamadas recursivas, correspondientes al cálculo de los dos números de fibonacci precedentes.

#### Implementación completa + ejecución:

````python
def fibonacci(n):
    """
    Calcula el término n-ésimo de la sucesión de Fibonacci.

    Parámetros:
        n (int): Posición en la sucesión, n >= 0.

    Retorna:
        int: Valor del término de Fibonacci en la posición n.
    """
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

if __name__ == "__main__":
    n_term = 6
    print(f"Término {n_term} de fibonacci: {fibonacci(n_term)}")
````

## 4. Número de cifras

Calcular el número de dígitos o cifras que conforman un número entero positivo $n \ge 0$ mediante una función recursiva.

#### 1. Análisis del problema y declaración del perfil de la función


```python
def num_cifras(n):
    """
    Calcula el número de cifras (dígitos) de un número entero.

    Parámetros:
        n (int): El número entero, n >= 0.

    Retorna:
        int: La cantidad total de dígitos que posee.
    """
```

#### 2. Análisis de casos

* **Caso base**: Si $n < 10$, se trata de un único dígito, por lo que el número de cifras será directamente 1.
* **Caso general**: Si $n \ge 10$, al dividirlo de forma entera por su base ($10$) le estaremos extrayendo su última cifra, es decir, el número de cifras en base $10$ se reduce. Por tanto, el número de cifras global del número $n$ estará compuesto por esa unidad que hemos abstraído en un salto ($1$) sumada a la cantidad de cifras del resto del número `num_cifras(n // 10)`.

#### 3. Transcripción

```python
if n < 10:
    return 1
else:
    return 1 + num_cifras(n // 10)
```

#### 4. Validación

- La llamada recursiva dividiendo $n$ entre $10$ reduce severamente y de forma inevitable y escalar la cifra tratada. A base de sustracción escalada (división entera), cualquier número elevado terminará cruzando la barrera hacia $n \le 9$, cumpliendo con el **caso base** y garantizando la finalización segura de la operación.

#### Tipo de recursión

**Recursión lineal no final**: al resultado del número de cifras del número con una cifra menos se le suma una unidad. 

#### Implementación completa + ejecución:

````python
def num_cifras(n):
    """
    Calcula el número de cifras (dígitos) de un número entero.
    
    Parámetros:
        n (int): Un número entero, n >= 0.
        
    Retorna:
        int: El total de dígitos.
    """
    if n < 10:
        return 1
    else:
        return 1 + num_cifras(n // 10)

if __name__ == "__main__":
    n_test = 23410
    print(f"Número de cifras de {n_test}: {num_cifras(n_test)}")
````

## 5. Suma de cifras

Calcula la suma de todas las cifras de un número entero positivo mediante una función recursiva.

#### 1. Análisis del problema y declaración del perfil de la función

```python
def suma_cifras(n):
    """
    Calcula la suma de todas las cifras de un número entero positivo.

    Parámetros:
        n (int): El número entero inicial en base 10, n >= 0.

    Retorna:
        int: Suma de todos los dígitos del número.
    """
```

#### 2. Análisis de casos

* **Caso base**: Si $n < 10$, el número solo tiene un dígito, por tanto su suma de cifras es el propio $n$.
* **Caso general**: Si $n$ tiene dos o más cifras ($n \ge 10$), el resultado será la suma de la última cifra con la suma de las cifras del número resultante tras dividirlo entre 10.

#### 3. Transcripción

```python
if n < 10:
    return n
else:
    return (n % 10) + suma_cifras(n // 10)
```

#### 4. Validación

- En cada llamada subsiguiente, $n$ se divide entre 10, acercándolo paulatamente a 0, y en particular, a un rango de valores inferior a 10, lo que garantiza su terminación.
- Dada la condición inicial $n \ge 0$, $n$ será positivo en todas las llamadas subsiguientes, ya que siempre se divide entre un número entero positivo (10).


#### Tipo de recursión

**Recursión lineal no final**: a la suma de cifras del problema más sencillo debemos sumarle la última cifra del número actual. 

#### Implementación completa + ejecución:

````python
def suma_cifras(n):
    """
    Calcula la suma de todas las cifras de un número entero positivo.

    Parámetros:
        n (int): El número entero inicial en base 10, n >= 0.

    Retorna:
        int: Suma de todos los dígitos del número.
    """
    if n < 10:
        return n
    else:
        return (n % 10) + suma_cifras(n // 10)

if __name__ == "__main__":
    cifra_n = 234
    print(f"Suma de cifras de {cifra_n}: {suma_cifras(cifra_n)}")
````
