# T3.1 Recursión: conceptos, tipología, pila de llamadas y diseño

## 1. Introducción
Se denomina recursivo a cualquier ente (definición, proceso, estructura, etc.) que se define en función de sí mismo.

En programación, una función es recursiva si su resolución requiere la solución previa de la función para un caso más sencillo. Es decir, ¡si se invoca a sí misma!

**Ejemplo: Función factorial**
$$
\text{factorial}(n) = \begin{cases} 
1 & \text{si } n = 0 \\
n \cdot \text{factorial}(n-1) & \text{si } n > 0 
\end{cases}
$$

Para resolver un problema complejo mediante un algoritmo recursivo, se descompone el problema en una serie de problemas más simples que se resuelven empleando el mismo algoritmo, para luego componer la solución del problema complejo en base a las soluciones de los problemas más simples.

Hay dos elementos vitales en la recursión (estructura dual):
* **Caso directo o base**: el problema es lo suficientemente simple para ser resuelto de manera trivial, sin realizar más llamadas.
* **Caso recursivo o general**: la resolución del problema viene expresada en términos de la resolución de uno o más subproblemas del mismo tipo y más pequeños en alguna medida.

Una vez planteado un algoritmo recursivo es necesario verificar:
* **Terminación**: que en algún momento se llegará al caso base a partir de cualquier caso general planteado inicialmente.
* **Corrección**: que para cualquier subproblema que se dé, la solución del algoritmo es correcta (inducción matemática).

## 2. Estructura de un algoritmo recursivo

La estructura fundamental de un algoritmo recursivo consiste en tener una instrucción condicional (`if/else`) que permita distinguir entre el caso base y el caso general.

Siguiendo con el ejemplo del factorial:

````python
def factorial(n):
    if n == 0: # Caso base                
        return 1  
    else:      # Caso general
        return n * factorial(n - 1)  

print(f'El factorial de 3 es: {factorial(3)}')
````

Los problemas recursivos, como vemos, requieren repetir una serie de comprobaciones y operaciones para resolverlos. Por tanto, tienen una **naturaleza repetitiva**. 

Por ello, es posible resolver un problema recursivo utilizando una aproximación iterativa equivalente, mediante el uso de estructuras de control de repetición, como bucles `for` o `while`:

````python
def factorial_iterativo_for(n):
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

def factorial_iterativo_while(n):
    resultado = 1
    i = 1
    while i <= n:
        resultado *= i
        i += 1
    return resultado

def factorial_iterativo_while_reverse(n):
    resultado = 1
    i = n
    while i >= 1:
        resultado *= i
        i -= 1
    return resultado

print(f'El factorial iterativo (for) de 3 es: {factorial_iterativo_for(3)}')
print(f'El factorial iterativo (while) de 3 es: {factorial_iterativo_while(3)}')
print(f'El factorial iterativo (while con recorrido inverso) de 3 es: {factorial_iterativo_while_reverse(3)}')
````

No obstante, para el caso concreto de factorial, la solución recursiva es mucho más **natural** (leíble, entendible) que la iterativa. 

_(De hecho, la implementación recursiva es una transcripción literal de la formulación formal matemática)_

En general, y aunque discutiremos los matices más adelante en el tema, **preferiremos las soluciones más "naturales" al problema**, en nuestra apuesta clara por **generar código legible y fácil de mantener**. 

> **Aviso Importante**: Como regla general, ¡**los algoritmos recursivos NO usan bucles** (`for` o `while`)! 
> 
> El propósito fundamental de la recursión es explotar la pila de memoria del sistema (como ilustraremos más adelante) como mecanismo de repetición de instrucciones, sustituyendo el uso de estructuras de control de repetición explícitas como `for` o `while`.
>
> Por tanto, como regla general, cuando implementes algoritmos recursivos: **¡BUCLES NO!** (salvo cuando realmente corresponda)

## 3. Tipos de recursión
* **Recursión Lineal**: hay a lo sumo una sola llamada recursiva en cada ejecución del algoritmo.
  * **Final**: el resultado de la llamada recursiva es el propio resultado de la llamada actual; es decir, la solución del caso más simple es a su vez la del caso más complejo.
  * **No final**: el resultado de la llamada recursiva se emplea para calcular el resultado de la llamada actual, arrojando un resultado posiblemente distinto al de la llamada recursiva. 
* **Recursión Múltiple**: hay más de una llamada recursiva en cada ejecución del algoritmo, y sus resultados han de combinarse. Se genera un árbol de llamadas.

Por ejemplo, `factorial(n)` es una función recursiva **lineal no final**, ya que el resultado de la llamada recursiva es multiplicado por `n`. Este es el esquema de la secuencia de llamadas que genera `factorial(3)`:

```text
factorial(3)
│
├──> factorial(2)
│   │
│   ├──> factorial(1)
│   │   │
│   │   ├──> factorial(0) ────> retorna 1
│   │   │
│   │   └─────────────────────> retorna 1 * 1 = 1
│   │
│   └─────────────────────────> retorna 2 * 1 = 2
│
└─────────────────────────────> retorna 3 * 2 = 6
```

A lo largo del tema iremos viendo diferentes tipos de recursión.

## 4. Pila de llamadas y recursividad

Una llamada a una función está asociada a un **registro de activación** (o *stack frame*) propio.

Un **registro de activación** almacena el **contexto** o estado de la llamada a función asociada. En concreto:
- almacena copias de variables locales de la llamada a función correspondiente, 
- almacena metainformación relativa al flujo de ejecución: en qué punto (en qué instrucción) debería retomarse la función cuando el flujo de ejecución retorne a la misma. 

Dichos registros se apilan en la **pila de llamadas** (*call stack*). 
- En el tope de la pila se ubica el registro de activación de la función que se está ejecutando en el momento actual. 
- Debajo del tope se encuentran registros de activación que estan esperando a que terminen los registros de activación superiores para proseguir su ejecución. 

Cuando una ejecución de un método finaliza, deja de existir su registro de activación (se desapila), y la ejecución se reanuda en la función que lo llamó, recuperando su propio contexto.

Los algoritmos recursivos hacen un uso muy intensivo de la pila de llamadas, ya que cada llamada de una función a sí mismo genera un nuevo registro de activación. 

### Ejemplo: pila de llamadas para factorial

Veamos el código completo de `factorial()` y una función `main()` invocándola, para luego visualizar cómo esto se traduce en la gestión de la memoria del ordenador (pila de llamadas):

````python
def factorial(n):
    if n == 0:
        ret = 1
    else:
        ret = n * factorial(n - 1)
    return ret

def main():
    print("Inicio del programa main()")
    resultado = factorial(3)
    print(f"El factorial de 3 es: {resultado}")
    print("Fin del programa")

main()
````

### Límites de la recursividad

En Python, existe un límite de llamadas recursivas para evitar que se desborde la pila (*Stack Overflow*). Podemos consultar e incluso cambiar este límite en Python mediante el módulo `sys`.

````python
import sys

print('Límite de recursión actual en Python:', sys.getrecursionlimit())

#sys.setrecursionlimit(1000) # podemos modificar el límite de recursión
````

Si excedemos el límite de llamadas recursivas pendientes de terminación, el intérprete lanza una excepción `RecursionError`. 

Vamos a probarlo realizando una recursión potencialmente infinita:

````python
def recursion_infinita(n=1):
    return recursion_infinita(n+1)

import inspect # paquete inspect, nos permite acceder a la traza y a los registros de activación

try:
    recursion_infinita()
except RecursionError:
    n_max = inspect.trace()[-1][0].f_locals.get('n') # Obtenemos el valor de n en el último registro
    print(f"¡Stack overflow! La pila se ha desbordado, con n = {n_max}.")
````

Como vemos, `n` no ha llegado al máximo. ¿Por qué? 

Pues porque la pila ya está ocupada por diversos registros de activación (wrappers) que introduce tu IDE (Jupyter Notebook, VSCode, etc.):

````python
def inspeccionar_pila():
    pila_actual = inspect.stack()
    print(f"La pila tiene {len(pila_actual)} registros activos actualmente.")
    
    i = len(pila_actual) - 1
    for registro in pila_actual:
        print(f"Nivel {i}: {registro.function} en {registro.filename}")
        i -= 1

inspeccionar_pila()
````

## 5. Etapas del diseño recursivo

El diseño recursivo consta de cuatro etapas:

#### 1. **Análisis del problema y declaración del perfil de la función**: 

Entender el problema y establecer el enunciado formal:
- Declarar la cabecera o perfil de la función (nombre de la función, parámetros, retorno).
- Establecer las condiciones de los parámetros de entrada (restricciones).
- Establecer el resultado esperado.

Las condiciones y el resultado se suelen **documentar** en un comentario multilinea tipo [**docstring**](https://peps.python.org/pep-0257/) dentro de la función.

#### 2. **Análisis de casos**: 

Identificar conceptualmente el caso base y el caso general, y establecer las acciones a ejecutar en cada caso.
- Notar que en ciertos problemas puede haber más de un caso base.
- Hay que comprobar que se cubre cualquier caso posible.

#### 3. **Transcripción**: 

Transcribir los casos y las acciones a realizar en forma de algoritmo recursivo en el lenguaje de programación deseado.

#### 4. **Validación**: 

Determinar que, en cada nueva llamada recursiva:
  + el nuevo problema a resolver es *estrictamente más cercano al caso base*.
  + las sucesivas llamadas recursivas cumplen con las condiciones de entrada.


### Ejemplo: diseño del algoritmo recursivo Factorial

1. **Análisis del problema y declaración del perfil de la función**: 

    ```python
    def factorial(n):
        """ Devuelve el factorial de un número entero no negativo.
    
        Parámetros:
            n (int): Un número entero no negativo (n >= 0).
    
        Retorna:
            int: El factorial de n.
        """
   ```
2. **Análisis de casos**: 
   - *Caso base*: `n == 0` $\rightarrow$ El factorial es 1.
   - *Caso general*: `n > 0` $\rightarrow$ El factorial es $n \cdot \text{factorial}(n-1)$.

3. **Transcripción**: 
    
    ```python
    if n == 0:
        ret = 1
    else:
        ret = n * factorial(n - 1)
    return ret
    ```

4. **Validación**: 
    - En la llamada `factorial(n-1)`, el parámetro es estrictamente menor, acercándose hacia 0. Como `n` es entero positivo, alcanzará obligatoriamente el caso base `0`, asegurando la terminación.
    - Dado que $n \ge 0$ en la llamada inicial, que en toda llamada subsiguiente $n$ se decrementa en una unidad, y que el caso base se da con $n = 0$, se garantiza que $n$ nunca será negativo. 

## 6. Resumen

En este cuaderno hemos introducido la **recursividad**, un concepto fundamental en la programación donde una función se llama a sí misma para resolver problemas descomponiéndolos en subproblemas más simples.

Los puntos clave son:
* **Estructura Dual**: Todo algoritmo recursivo válido debe poseer al menos un **caso base** (condición de parada trivial) y un **caso general** (llamada a sí mismo con un problema reducido).
* **Ausencia de Bucles**: La recursión sustituye el uso explícito de iteraciones (`for`, `while`) apoyándose en la infraestructura subyacente del lenguaje (pila de llamadas).
* **Tipos de recursión**: lineal (final / no final) o múltiple. 
* **Pila de Llamadas**: Cada invocación recursiva genera un nuevo marco o **registro de activación** en memoria, preservando el estado (variables locales) de cada llamada. Estos registros se apilan y se desapilan conforme se generan nuevas llamadas y estas terminan, respectivamente.
* **Diseño Metodológico**: Su desarrollo se aborda estructuradamente mediante la definición matemática, el análisis estricto de casos, la redacción en código y la validación matemática de su avance hacia el caso base.
