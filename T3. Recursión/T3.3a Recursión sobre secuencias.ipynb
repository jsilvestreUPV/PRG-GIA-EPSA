# T3.3a Recursión sobre secuencias: recorrido y búsqueda

Vamos a poner en práctica la recursión aplicada a estructuras de datos lineales (secuencias). En próximos cursos aplicaréis recursión sobre estructuras de datos más complejas (p.e. árboles) donde el uso de la recursión resulta mucho más natural y conveniente que con secuencias.

Los algoritmos que trabajan con secuencias (listas) se pueden formular de manera recursiva **reduciendo el problema a subproblemas que trabajen con sub-secuencias más pequeñas**, hasta alcanzar subsecuencias de tamaño unitario o nulo (lo que constituiría un caso base de resolución inmediata).

Ejemplo: para sumar los elementos de una lista de enteros, el problema se puede reformular recursivamente como la suma de:
- el primer elemento de la lista, 
- más la suma de elementos de la lista resultante de eliminar el primer elemento.

## 1. Eficiencia computacional para trabajar con subsecuencias

En Python, parece natural usar _slices_ del listas para generar las sub-secuencias de las que depende la recursión:

````python
lista_original = [1, 2, 3, 4, 5]
print(lista_original)

lista2 = lista_original[1:] # slice: todos los elementos desde la posición 1
print(lista2)

lista3 = lista_original[:-1] # slice: todos los elementos excepto el último
print(lista3)

lista4 = lista_original[1:-1] # slice: todos los elementos excepto el primero y el último
print(lista4)
````

No obstante, **hacer *slicing* es tremendamente ineficiente en el contexto de la recursión**, ya que un *slice* genera **una copia nueva** en memoria de la subsecuencia en cada llamada recursiva. 

Por ello, en el contexto de la recursión, a fin de preservar coste temporal y computacional (eficiencia), aplicaremos la técnica general (común a lenguajes como C++ y Java) de trabajar **sobre la lista original** pero usando variables índice que delimiten las posiciones en las que empieza o acaba una subsecuencia.


````python
lista_original = [1, 2, 3, 4, 5]

print("Trabajando con slices:")
lista_sliced = lista_original[1:-1] # esto crea una nueva lista en memoria!
for elem in lista_sliced:
    print(elem)

print("\nTrabajando con índices (más eficiente):")
beg = 1; end = 3  # delimitamos la subsecuencia con las posiciones de inicio y fin
for i in range(beg, end+1):
    print(lista_original[i])
````

Por convención:

- Llamaremos `seq` a las secuencias, que pueden ser indistintamente listas o tuplas de Python. 
- La variables índice las llamaremos:
  - `beg` (_begin_, inicio): representa el límite izquierdo o inferior de la subsecuencia. 
  - `end` (fin): representa el límite derecho o superior de la subsecuencia. 
- Dada una secuencia `seq`, nos referiremos con `seq[beg..end]` a la subsecuencia que va desde la posición `beg` a la `end` (ambas incluidas) de la secuencia original `seq`.
- El tamaño de la subsecuencia será `end - beg + 1`. 
- Cuando `beg > end`, se considerará que la secuencia es vacía o de tamaño cero.

La figura inferior muestra una secuencia `seq` de 8 elementos, donde se ha sombreado la subsecuencia `seq[beg..end]` delimitada por los índices `beg=2` y `end=5`.

## 2. Recorrido de secuencias

### 2.1 Consideraciones generales

En un recorrido de una secuencia `seq`, típicamente se deben tratar todas las componentes de `seq[beg..end]`. Su resolución recursiva suele contemplar el siguiente análisis de casos:

- **Caso base**: La subsecuencia `seq[beg..end]` es suficientemente pequeña como para que la solución sea directa. Dependiendo del problema, suele ser:
    - `beg == end`: Subsecuencia con una sola componente.
    - `beg > end`: Subsecuencia vacía.
    - `beg >= end`: Subsecuencia con a lo sumo una componente.
- **Caso general**: Secuencias más grandes donde el recorrido se resuelve con un subrecorrido sobre una parte de la secuencia más pequeña:
    - `seq[beg + 1..end]` para un recorrido ascendente
       - `end` es siempre el mismo valor, típicamente `len(seq)`.
    - `seq[beg..end-1]` para un recorrido descendende.
       - `beg` es siempre el mismo valor, típicamente `0`.
    - `seq[beg + 1..end -1]` para un recorrido combinado.

La secuencia y los índices son **parámetros de la función recursiva**.
- Los índices que son invariables pueden omitirse de la lista de parámetros, si queremos.

### 2.2 Problemas


### 2.2.1 Suma de las componentes de una secuencia

Dada una secuencia `seq` de valores numéricos, calcular la suma de sus componentes, realizando un recorrido ascendente.

#### 1. Análisis del problema y declaración del perfil de la función

```python
def suma(seq, beg, end):
    """
    Calcula la suma de los elementos (numéricos) de una secuencia desde beg hasta end.

    Parámetros:
        seq: Secuencia de valores númericos.
        beg (int): Posicion inicial del recorrido en seq. Condición: beg >= 0.
        end (int): Posición final del recorrido en seq. Condición: end < len(seq).

    Retorna:
        int/float: Suma de los elementos de seq desde beg hasta end.
    """
```

#### 2. Análisis de casos

Aquí variaremos `beg` para hacer el recorrido ascendente. 

* **Caso base**: Si el índice `beg` ha superado el límite de la subsecuencia (`beg > end`), la subsecuencia es vacía, por tanto, la suma es 0.
* **Caso general**: En otro caso, la suma es el valor del elemento actual `seq[beg]` más la suma de la subsecuencia `seq[beg+1..end]`.

#### 3. Transcripción

```python
if beg > end:
    return 0
else:
    return seq[beg] + suma(seq, beg + 1, end)
```

#### 4. Validación

- En cada llamada recursiva incrementamos `beg` en una unidad, garantizando que eventualmente alcanzaremos el caso base.

#### Implementación completa + ejecución:

````python
def suma(seq, beg, end):
    """
    Calcula la suma de los elementos (numéricos) de una secuencia desde beg hasta end.

    Parámetros:
        seq: Secuencia de valores númericos.
        beg (int): Posicion inicial del recorrido en seq. Condición: beg >= 0.
        end (int): Posición final del recorrido en seq. Condición: end < len(seq).

    Retorna:
        int/float: Suma de los elementos de seq desde beg hasta end.
    """
    if beg > end:
        return 0
    else:
        return seq[beg] + suma(seq, beg + 1, end)

if __name__ == "__main__":
    lista = [2.0, 3.5, 4.0, -1.0, 2.5]
    suma_total = suma(lista, 0, len(lista)-1)
    print('Suma de elementos:', suma_total)
````

#### Optimización: 

Eliminamos `end` (siempre es `len(seq)-1`!), y nos ahorramos almacenar una variable (un puntero) por cada registro de activación.

````python
def suma(seq, beg):
    """
    Calcula la suma de los elementos de una secuencia desde beg hasta len(seq)-1.

    Parámetros:
        seq (list/tuple): Secuencia de valores númericos.
        beg (int): Posicion inicial del recorrido en seq. Condición: beg >= 0.
        
    Retorna:
        int/float: Suma de los elementos.
    """
    if beg >= len(seq):
        return 0
    else:
        return seq[beg] + suma(seq, beg + 1)

if __name__ == "__main__":
    lista = [2.0, 3.5, 4.0, -1.0, 2.5]
    suma_total = suma(lista, 0)
    print('Suma de elementos:', suma_total)
````

### 2.2.2 Posición del mínimo de una secuencia

Devolver la posición (el índice) donde se encuentra el mínimo elemento de la secuencia, realizando un recorrido descendiente.

#### 1. Análisis del problema y declaración del perfil de la función

```python
def minimo(seq, beg, end):
    """
    Devuelve el índice de la posición del mínimo elemento de la secuencia, realizando un recorrido descendente.

    Parámetros:
        seq: Secuencia de objetos comparables.
        beg (int): Posición inicial del recorrido en seq. Condición: beg >= 0.
        end (int): Posición final del recorrido en seq. Condición: end < len(seq).

    Retorna:
        int: El índice donde se encuentra el elemento mínimo.
    """
```

#### 2. Análisis de casos

Aquí variaremos `end` para hacer el recorrido descendente. 

* **Caso base**: Si se considera una subsecuencia de un solo elemento (`end == beg`), ese elemento es el mínimo y devolveremos su posición.
* **Caso general**: Si no, la subsecuencia tiene 2 o más elementos, por tanto, compararemos el valor mínimo del resto de la secuencia `seq[beg..end-1]` com el elemento actual `seq[end]`, devolviendo la posición del mínimo de ambos. 

#### 3. Transcripción

```python
if end == beg:
    return end
else:
    pos_min = minimo(seq, beg, end - 1)
    if seq[end] < seq[pos_min]:
        return end
    else:
        return pos_min
```

#### 4. Validación

- Al igual que en el caso anterior, el decremento unitario de `end` garantiza la terminación al acercarse al principio de la secuencia.

#### Implementación completa + ejecución:

````python
def minimo(seq, beg, end):
    """
    Devuelve el índice de la posición del mínimo elemento de la secuencia, realizando un recorrido descendente.

    Parámetros:
        seq: Secuencia de objetos comparables.
        beg (int): Posición inicial del recorrido en seq. Condición: beg >= 0.
        end (int): Posición final del recorrido en seq. Condición: end < len(seq).

    Retorna:
        int: El índice donde se encuentra el elemento mínimo.
    """
    if end == beg:
        return end
    else:
        pos_min = minimo(seq, beg, end - 1)
        if seq[end] < seq[pos_min]:
            return end
        else:
            return pos_min

if __name__ == "__main__":
    lista = [1.5, 4.0, -1.0, 1.0, 2.5]
    print('La posición del mínimo es:', minimo(lista, 0, len(lista)-1))
````

#### Optimización: 

Eliminamos `beg` (siempre es 0!), y nos ahorramos almacenar una variable (un puntero) por cada registro de activación.

````python
def minimo(seq, end):
    """
    Devuelve el índice de la posición del valor mínimo en la secuencia.

    Parámetros:
        seq: Secuencia de elementos comparables.
        end (int): Posición final del recorrido en seq. Condición: end < len(seq).

    Retorna:
        int: El índice donde se encuentra el valor mínimo.
    """
    if end == 0:
        return end
    else:
        pos_min = minimo(seq, end - 1)
        if seq[end] < seq[pos_min]:
            return end
        else:
            return pos_min

if __name__ == "__main__":
    lista = [1.5, 4.0, -1.0, 1.0, 2.5]
    print('La posición del mínimo es:', minimo(lista, len(lista)-1))
````

### 2.2.3 Invertir las componentes de una secuencia mutable

Dada una secuencia mutable `seq`, invertir el orden de sus componentes. Por ejemplo: `[5, 2, 1, 4, 3]` se convierte en `[3, 4, 1, 2, 5]`.

#### 1. Análisis del problema y declaración del perfil de la función

```python
def invertir(seq, beg, end):
    """
    Invierte los elementos de seq[beg..end].

    Parámetros:
        seq: Secuencia original. Debe ser mutable (list).
        beg (int): Posición inicial del recorrido en seq. Condición: beg >= 0.
        end (int): Posición final del recorrido en seq. Condición: end < len(seq).
    Retorna:
        None (los cambios en la lista se realizan in-place sobre seq)
    """
```

#### 2. Análisis de casos

* **Caso base**: Si la subsecuencia tiene una o cero componentes, no hay nada que invertir. No hacemos nada (caso base "vacío").
* **Caso general**: Si la subsecuencia tiene más de una componente, intercambiamos el primer elemento `beg` con el último `end` y llamamos recursivamente a la función para el resto de la subsecuencia `seq[beg+1..end-1]`.

#### 3. Transcripción

```python
    if beg < end:
        aux = seq[beg]
        seq[beg] = seq[end]
        seq[end] = aux
        invertir(seq, beg + 1, end - 1)
```

#### 4. Validación

- En cada llamada recursiva decrementamos `end`, por lo que la distancia `end - beg` disminuye, garantizando que alcancemos el caso base `beg > end`.

#### Implementación completa + ejecución:

````python
def invertir(seq, beg, end):
    """
    Invierte los elementos de seq[beg..end].

    Parámetros:
        seq: Secuencia original. Debe ser mutable (list).
        beg (int): Posición inicial del recorrido en seq. Condición: beg >= 0.
        end (int): Posición final del recorrido en seq. Condición: end < len(seq).
    Retorna:
        None (los cambios en la lista se realizan in-place sobre seq)
    """
    if beg < end:
        aux = seq[beg]
        seq[beg] = seq[end]
        seq[end] = aux
        invertir(seq, beg + 1, end - 1)

if __name__ == "__main__":   
    lista = [1, 2, 3, 4, 5]
    print(lista)
    invertir(lista, 0, len(lista) - 1)
    print(lista)
````

## 3. Búsqueda

Al igual que el recorrido, una búsqueda recursiva reduce su espacio de búsqueda comprobando un elemento y prosiguiendo la búsqueda en una subsecuencia más pequeña.

La búsqueda **terminará prematuramente** en cuanto tenga éxito sin necesidad de procesar toda la secuencia.

### 3.1 Consideraciones generales

En una búsqueda recursiva, la zona de búsqueda se va reduciendo a subsecuencias más pequeñas. 

La recursión finaliza:
- cuando se encuentra lo que se busca (respuesta), o 
- cuando se agotan los componentes de la secuencia, ya que la búsqueda no puede tener éxito.

El análisis de casos sigue este esquema:

- **Caso base**: Subsecuencia reducida al tamaño más pequeño posible, donde podemos responder directamente, o bien donde la búsqueda **no puede tener éxito** (p.e., `beg > end`).
- **Caso general**: La subsecuencia tiene suficientes componentes para proseguir la búsqueda. Entonces:
    - **Éxito**: Si se encuentra lo que se busca, la búsqueda termina inmediatamente con independencia de los elementos que queden por revisar.
    - **No éxito**: Si no se encuentra lo que se busca, se reduce la búsqueda a una subsecuencia con menos componentes.

### 3.2 Problemas


### 3.2.1 Búsqueda Secuencial

Devolver el índice de la posición en la que se encuentra un objeto `x` en la secuencia `seq`, realizando un recorrido ascendente. Si no lo encuentra, devuelve `-1`.

#### 1. Análisis del problema y declaración del perfil de la función

```python
def buscar(seq, beg, x):
    """
    Busca de forma secuencial el objeto (valor) x en la secuencia seq desde el índice beg.

    Parámetros:
        seq: Secuencia de elementos.
        beg (int): Posicion inicial del recorrido en seq. Condición: beg >= 0.
        x: Objeto (valor) a buscar.

    Retorna:
        (int): El índice de x si se encuentra en la secuencia, -1 en caso contrario.
    """
```

#### 2. Análisis de casos

* **Casos base**:
    * Si el índice inicial ha superado el tamaño de la secuencia (`beg >= len(seq)`), la subsecuencia es vacía. No se encuentra el valor, y devolvemos -1.
    * Si el elemento actual coincide con el buscado (`seq[beg] == x`), hemos tenido éxito y devolvemos la posición actual `beg`.
* **Caso general**: En otro caso, buscamos recursivamente en el resto de la secuencia `seq[beg + 1...]`.

#### 3. Transcripción

```python
if beg >= len(seq):
    return -1
elif seq[beg] == x:
    return beg
else:
    return buscar_seq(seq, beg + 1, x)
```

#### 4. Validación

- El incremento unitario de `beg` asegura que eventualmente alcanzaremos el final de la secuencia o encontraremos el elemento, garantizando la terminación.

#### Implementación completa + ejecución:

````python
def buscar(seq, beg, x):
    """
    Busca de forma secuencial el objeto (valor) x en la secuencia seq desde el índice beg.

    Parámetros:
        seq: Secuencia de elementos.
        beg (int): Posicion inicial del recorrido en seq. Condición: beg >= 0.
        x: Objeto (valor) a buscar.

    Retorna:
        (int): El índice de x si se encuentra en la secuencia, -1 en caso contrario.
    """
    if beg >= len(seq):
        return -1
    elif seq[beg] == x:
        return beg
    else:
        return buscar(seq, beg + 1, x)

if __name__ == "__main__":
    lista = [4, 6, 9, 11, 34, 65, 82]
    print(lista)
    print('La posición de 34 es:', buscar(lista, 0, 34))
    print('La posición de 5 es:', buscar(lista, 0, 5))
````

### 3.2.2 Comprobar si una secuencia está ordenada

Comprobar si una secuencia está ordenada ascendentemente.

#### 1. Análisis del problema y declaración del perfil de la función

```python
def ordenada(seq, beg):
    """
    Comprueba si la secuencia seq está ordenada de forma ascendente desde beg.

    Parámetros:
        seq: Secuencia de elementos comparables.
        beg (int): Posicion inicial del recorrido en seq. Condición: beg >= 0.

    Retorna:
        bool: True si está ordenada, False en caso contrario.
    """
```

#### 2. Análisis de casos

Enfocaremos el algoritmo como la búsqueda de una pareja de elementos que no estén ordenados.

* **Casos base**:
    * Si la subsecuencia tiene uno o ningún elemento (`beg >= len(seq) - 1`), se considera ordenada por defecto, devolvemos `True`.
    * Si la subsecuencia tiene dos o más elementos, y el elemento actual `seq[beg]` es mayor que el siguiente `seq[beg + 1]`, la subsecuencia no está ordenada, por lo que devolvemos `False` inmediatamente (detección de fallo en la búsqueda).
* **Caso general**: En otro caso, comprobamos recursivamente el resto de la secuencia `seq[beg + 1..])`.

#### 3. Transcripción

```python
if beg >= len(seq) - 1:
    return True
else:
    if seq[beg] > seq[beg + 1]:
        return False
    else:
        return ordenada(seq, beg + 1)
```

#### 4. Validación

- El incremento unitario de `beg` asegura que eventualmente alcanzaremos el final de la secuencia o encontraremos un par desordenado, garantizando la terminación.

#### Implementación completa + ejecución:

````python
def ordenada(seq, beg):
    """
    Comprueba si la secuencia seq está ordenada de forma ascendente desde beg.

    Parámetros:
        seq: Secuencia de elementos comparables.
        beg (int): Posicion inicial del recorrido en seq. Condición: beg >= 0.

    Retorna:
        bool: True si está ordenada, False en caso contrario.
    """
    if beg >= len(seq) - 1:
        return True
    else:
        if seq[beg] > seq[beg + 1]:
            return False
        else:
            return ordenada(seq, beg + 1)

if __name__ == "__main__":
    lista = [2, 3, 5, 2]
    print(f'¿{lista} está ordenada?:', ordenada(lista, 0))
    lista = [2, 3, 5, 9]
    print(f'¿{lista} está ordenada?:', ordenada(lista, 0))
````

### 3.2.3 Búsqueda Binaria (*Dicotómica*)

Devolver el índice de la posición en la que se encuentra un objeto `x` en una secuencia `seq` **ordenada**. Si no lo encuentra, devuelve `-1`.

Notar que, cuando sabemos de antemano que la secuencia `seq` **está ordenada**, podemos descartar la mitad de los elementos en cada paso comparando el valor `x` con el elemento central de la subsecuencia `seq`.

#### 1. Análisis del problema y declaración del perfil de la función

```python
def busqueda_binaria(seq, beg, end, x):
    """
    Busca binaria del objeto (valor) x en la secuencia seq ordenada ascendentemente, desde beg hasta end.

    Parámetros:
        seq: Secuencia de elementos comparables. Debe estar ordenada ascendentemente.
        beg (int): Índice inferior de la subsecuencia. Condición: beg >= 0.
        end (int): Índice superior de la subsecuencia. Condición: end < len(seq).
        x: Objeto (valor) a buscar.

    Retorna:
        (int): El índice de x si se encuentra en la secuencia, -1 en caso contrario.
    """
```

#### 2. Análisis de casos

* **Casos base**:
    * Si la subsecuencia es vacía (cero elementos, `beg > end`), la búsqueda no puede tener éxito y devolvemos -1.
    * Si el elemento en la posición media de la subsecuencia es igual a `x`, hemos tenido éxito y devolvemos dicha posición.
* **Caso general**: Si `x` es mayor que `seq[medio]`, buscamos recursivamente en la mitad superior `seq[medio + 1..end]`. Si es menor, buscamos en la mitad inferior `seq[beg..medio - 1]`.

#### 3. Transcripción

```python
if beg > end:
    return -1
else:
    medio = (beg + end) // 2
    if seq[medio] == x:
        return medio
    elif seq[medio] < x:
        return busqueda_binaria(seq, medio + 1, end, x)
    else:
        return busqueda_binaria(seq, beg, medio - 1, x)
```

#### 4. Validación

- En cada paso, el espacio de búsqueda se reduce a la mitad, garantizando que eventualmente la subsecuencia sea vacía o que se encuentre el elemento, garantizando la terminación.

#### Implementación completa + ejecución:

````python
def busqueda_binaria(seq, beg, end, x):
    """
    Busca binaria del objeto (valor) x en la secuencia seq ordenada ascendentemente, desde beg hasta end.

    Parámetros:
        seq: Secuencia de elementos comparables. Debe estar ordenada ascendentemente.
        beg (int): Índice inferior de la subsecuencia. Condición: beg >= 0.
        end (int): Índice superior de la subsecuencia. Condición: end < len(seq).
        x: Objeto (valor) a buscar.

    Retorna:
        (int): El índice de x si se encuentra en la secuencia, -1 en caso contrario.
    """
    if beg > end:
        return -1
    else:
        medio = (beg + end) // 2
        if seq[medio] == x:
            return medio
        elif seq[medio] < x:
            return busqueda_binaria(seq, medio + 1, end, x)
        else:
            return busqueda_binaria(seq, beg, medio - 1, x)

if __name__ == "__main__":
    lista_ord = [4, 6, 9, 11, 34, 65, 82]
    print(lista_ord)
    print('La posición de 11 es:', busqueda_binaria(lista_ord, 0, len(lista_ord) - 1, 11))
    print('La posición de 5 es:', busqueda_binaria(lista_ord, 0, len(lista_ord) - 1, 5))
````
