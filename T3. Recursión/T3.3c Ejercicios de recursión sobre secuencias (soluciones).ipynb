## 1. Duplicar los valores de una secuencia

Dada una secuencia mutable `seq`, duplicar el valor de todas sus componentes numéricas.

### 1. Análisis del problema y declaración del perfil de la función

```python
def duplicar(seq, beg):
    """
    Duplica el valor de los elementos de seq desde el índice beg hasta el final.

    Parámetros:
        seq (list): Secuencia de valores numéricos. Debe ser mutable.
        beg (int): Posición inicial del recorrido en seq. Condición: beg >= 0.
    Retorna:
        None: Los cambios se realizan in-place sobre seq.
    """
```

### 2. Análisis de casos

* **Caso base**: Si el índice `beg` ha alcanzado o superado el tamaño de la secuencia (`beg >= len(seq)`), la subsecuencia es vacía y no hay nada que duplicar.
* **Caso general**: En otro caso, duplicamos el elemento actual `seq[beg]` y llamamos recursivamente a la función para procesar el resto de la secuencia `seq[beg+1..]`.

### 3. Transcripción

```python
if beg < len(seq):
    seq[beg] *= 2
    duplicar(seq, beg + 1)
```

### 4. Validación

- El incremento unitario de `beg` en cada llamada garantiza que eventualmente alcanzará el tamaño de la secuencia (`len(seq)`), alcanzando el caso base y asegurando la terminación del algoritmo.

#### Implementación completa + ejecución:

````python
def duplicar(seq, beg):
    """
    Duplica el valor de los elementos de seq desde el índice beg hasta el final.

    Parámetros:
        seq (list): Secuencia de valores numéricos. Debe ser mutable.
        beg (int): Posición inicial del recorrido en seq. Condición: beg >= 0.
    Retorna:
        None: Los cambios se realizan in-place sobre seq.
    """
    if beg < len(seq):
        seq[beg] *= 2
        duplicar(seq, beg + 1)

if __name__ == "__main__":
    lista = [1, 2, 3, 4, 5]
    print("Original:", lista)
    duplicar(lista, 0)
    print("Duplicados:", lista)
````

## 2. Contar valores númericos dentro de un rango

Contar el número de valores numéricos (enteros, floats) de la secuencia cuyo valor esté dentro de un rango inclusivo `[min_val, max_val]`.

### 1. Análisis del problema y declaración del perfil de la función

```python
def contar_en_rango(seq, beg, min_val, max_val):
    """
    Cuenta cuántos elementos de seq desde el índice beg hasta el final están en el rango [min_val, max_val].

    Parámetros:
        seq: Secuencia de valores numéricos.
        beg (int): Posición inicial. Condición: beg >= 0.
        min_val: Valor mínimo del rango.
        max_val: Valor máximo del rango.
    Retorna:
        int: Cantidad de elementos dentro del rango.
    """
```

### 2. Análisis de casos

* **Caso base**: Si el índice `beg` ha alcanzado el final de la secuencia (`beg >= len(seq)`), la subsecuencia está vacía, y por tanto, el número de elementos que cumplen la condición son 0.
* **Caso general**: 
    * Si el elemento actual `seq[beg]` está en el rango, el resultado es 1 más el el número de elementos que estan dentro del rango de valores en la subsecuencia `seq[beg+1..]`.
    * Si no, el resultado es simplemente el número de elementos que estan dentro del rango de valores en la subsecuencia `seq[beg+1..]`.

### 3. Transcripción

```python
if beg >= len(seq):
    return 0
else:
    es_en_rango = 1 if min_val <= seq[beg] <= max_val else 0
    return es_en_rango + contar_en_rango(seq, beg + 1, min_val, max_val)
```

### 4. Validación

- El incremento de `beg` asegura que eventualmente alcanzaremos el caso base `beg >= len(seq)`.

#### Implementación completa + ejecución:

````python
def contar_en_rango(seq, beg, min_val, max_val):
    """
    Cuenta cuántos elementos de seq desde el índice beg hasta el final están en el rango [min_val, max_val].

    Parámetros:
        seq: Secuencia de valores numéricos.
        beg (int): Posición inicial. Condición: beg >= 0.
        min_val: Valor mínimo del rango.
        max_val: Valor máximo del rango.
    Retorna:
        int: Cantidad de elementos dentro del rango.
    """
    if beg >= len(seq):
        return 0
    else:
        es_en_rango = 1 if min_val <= seq[beg] <= max_val else 0
        return es_en_rango + contar_en_rango(seq, beg + 1, min_val, max_val)

if __name__ == "__main__":
    lista = [10, 5, 20, 15, 30]
    min_r, max_r = 10, 20
    print(f"Elementos en rango [{min_r}, {max_r}] en {lista}:")
    print(contar_en_rango(lista, 0, min_r, max_r))
````

## 3. Contar el número de ocurrencias de un objeto x

Contar cuántas veces aparece un elemento `x` en la secuencia.

### 1. Análisis del problema y declaración del perfil de la función

```python
def contar_ocurrencias(seq, beg, x):
    """
    Cuenta las ocurrencias de x en seq desde el índice beg hasta el final.

    Parámetros:
        seq: Secuencia de elementos.
        beg (int): Posición inicial. Condición: beg >= 0.
        x: Objeto a buscar.
    Retorna:
        int: Número de veces que x aparece en la subsecuencia.
    """
```

### 2. Análisis de casos

* **Caso base**: Si el índice `beg` ha alcanzado el final de la secuencia (`beg >= len(seq)`), la subsecuencia está vacía y entonces `x` aparece 0 veces.
* **Caso general**: 
    * Si `seq[beg] == x`, el resultado es 1 más el número de veces que `x` aparece en la subsecuencia `seq[beg+1..]`.
    * Si no, el resultado es el número de veces que `x` aparece en la subsecuencia `seq[beg+1..]`.

### 3. Transcripción

```python
if beg >= len(seq):
    return 0
else:
    coincide = 1 if seq[beg] == x else 0
    return coincide + contar_ocurrencias(seq, beg + 1, x)
```

### 4. Validación

- Al igual que en los problemas anteriores, el incremento de `beg` garantiza la terminación.

#### Implementación completa + ejecución:

````python
def contar_ocurrencias(seq, beg, x):
    """
    Cuenta las ocurrencias de x en seq desde el índice beg hasta el final.

    Parámetros:
        seq: Secuencia de elementos.
        beg (int): Posición inicial. Condición: beg >= 0.
        x: Objeto a buscar.
    Retorna:
        int: Número de veces que x aparece en la subsecuencia.
    """
    if beg >= len(seq):
        return 0
    else:
        coincide = 1 if seq[beg] == x else 0
        return coincide + contar_ocurrencias(seq, beg + 1, x)

if __name__ == "__main__":
    lista = [1, 2, 3, 2, 4, 2, 5]
    valor = 2
    print(f"Ocurrencias de {valor} en {lista}:")
    print(contar_ocurrencias(lista, 0, valor))
````

## 4. Obtener la posición del primer valor numérico diferente de 0

Buscar el primer elemento de la secuencia que no sea cero y devolver su índice. Si todos son cero o la secuencia está vacía, devolver -1.

### 1. Análisis del problema y declaración del perfil de la función

```python
def posicion_primero_distinto_cero(seq, beg):
    """
    Devuelve el índice del primer elemento de seq desde el índice beg hasta el final que es distinto de 0.

    Parámetros:
        seq: Secuencia de valores numéricos.
        beg (int): Posición inicial de búsqueda. Condición: beg >= 0.
    Retorna:
        int: El índice del primer elemento != 0, o -1 si no se encuentra.
    """
```

### 2. Análisis de casos

* **Casos base**:
    * Si el índice `beg` ha alcanzado el final de la secuencia (`beg >= len(seq)`), la subsecuencia es vacía y por tanto devolvemos -1 (no encontrado).
    * Si el elemento actual `seq[beg]` es distinto de 0, hemos terminado: devolveremos su posición `beg`.
* **Caso general**: En otro caso (`seq[beg] == 0`), buscamos en el resto de la secuencia `seq[beg+1..]`.

### 3. Transcripción

```python
if beg >= len(seq):
    return -1
elif seq[beg] != 0:
    return beg
else:
    return posicion_primero_distinto_cero(seq, beg + 1)
```

### 4. Validación

- El incremento de `beg` asegura que eventualmente alcanzaremos el final de la secuencia o encontraremos un valor no nulo.

#### Implementación completa + ejecución:

````python
def posicion_primero_distinto_cero(seq, beg):
    """
    Devuelve el índice del primer elemento de seq desde el índice beg hasta el final que es distinto de 0.

    Parámetros:
        seq: Secuencia de valores numéricos.
        beg (int): Posición inicial de búsqueda. Condición: beg >= 0.
    Retorna:
        int: El índice del primer elemento != 0, o -1 si no se encuentra.
    """
    if beg >= len(seq):
        return -1
    elif seq[beg] != 0:
        return beg
    else:
        return posicion_primero_distinto_cero(seq, beg + 1)

if __name__ == "__main__":
    lista1 = [0, 0, 7, 0, 5]
    lista2 = [0, 0, 0]
    print(f"Posición del primero != 0 en {lista1}: {posicion_primero_distinto_cero(lista1, 0)}")
    print(f"Posición del primero != 0 en {lista2}: {posicion_primero_distinto_cero(lista2, 0)}")
````

## 5. Comprobar si la secuencia es capicúa

Determinar si una secuencia es igual leyéndola de izquierda a derecha que de derecha a izquierda.

### 1. Análisis del problema y declaración del perfil de la función

```python
def es_capicua(seq, beg, end):
    """
    Comprueba si la subsecuencia seq[beg..end] es capicúa.

    Parámetros:
        seq: Secuencia de elementos.
        beg (int): Índice izquierdo.
        end (int): Índice derecho.
    Retorna:
        bool: True si es capicúa, False en caso contrario.
    """
```

### 2. Análisis de casos

* **Casos base**:
    * Si la subsecuencia tiene cero o un elemento (`beg >= end`), es capicúa por definición, devolvemos `True`.
    * Si los elementos en los extremos `seq[beg]` y `seq[end]` son diferentes, no es capicúa, devolvemos `False` inmediatamente.
* **Caso general**: Si los extremos coinciden, la secuencia será capicúa si la subsecuencia `seq[beg+1..end-1]` también lo es.

### 3. Transcripción

```python
if beg >= end:
    return True
elif seq[beg] != seq[end]:
    return False
else:
    return es_capicua(seq, beg + 1, end - 1)
```

### 4. Validación

- En cada llamada, reducimos la distancia entre `beg` y `end` en dos unidades, por lo que eventualmente se cruzarán o se encontrarán, alcanzando el caso base.

#### Implementación completa + ejecución:

````python
def es_capicua(seq, beg, end):
    """
    Comprueba si la subsecuencia seq[beg..end] es capicúa.

    Parámetros:
        seq: Secuencia de elementos.
        beg (int): Índice izquierdo.
        end (int): Índice derecho.
    Retorna:
        bool: True si es capicúa, False en caso contrario.
    """
    if beg >= end:
        return True
    elif seq[beg] != seq[end]:
        return False
    else:
        return es_capicua(seq, beg + 1, end - 1)

if __name__ == "__main__":
    lista1 = [1, 2, 3, 2, 1]
    lista2 = [1, 2, 3, 4, 5]
    print(f"¿Es {lista1} capicúa?: {es_capicua(lista1, 0, len(lista1)-1)}")
    print(f"¿Es {lista2} capicúa?: {es_capicua(lista2, 0, len(lista2)-1)}")
````
