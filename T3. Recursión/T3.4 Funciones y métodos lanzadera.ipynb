# T3.4 Funciones y métodos lanzadera

Vamos a explorar brevemente el concepto de **funciones y métodos lanzadera** (*wrapper functions/methods*), fundamental en el diseño de algoritmos recursivos para mejorar su usabilidad y mantener la encapsulación.

## 1. Funciones lanzadera

Una función recursiva requiere parámetros adicionales para controlar el estado de la recursión (como índices de inicio y fin en un recorrido o búsqueda). 

Sin embargo, el usuario final del código no debería verse obligado a proporcionar estos valores técnicos iniciales.

Una **función lanzadera** es una función envoltorio (*wrapper function*) sencilla que llama a la función recursiva con los valores iniciales adecuados de los parámetros que controlan la recursión.

### Ejemplo: Búsqueda Binaria

Como vimos anteriormente, la búsqueda binaria sobre una secuencia requiere delimitar el rango de búsqueda con los índices `beg` y `end`.

```python
def busqueda_binaria_recursiva(seq, beg, end, x):
    if beg > end:
        return -1
    else:
        medio = (beg + end) // 2
        if seq[medio] == x:
            return medio
        elif seq[medio] < x:
            return busqueda_binaria_recursiva(seq, medio + 1, end, x)
        else:
            return busqueda_binaria_recursiva(seq, beg, medio - 1, x)

def busqueda_binaria(seq, x):
    """
    Función LANZADERA para la búsqueda binaria.
    Proporciona una interfaz limpia al usuario.
    """
    # La lanzadera se encarga de calcular los índices iniciales
    return busqueda_binaria_recursiva(seq, 0, len(seq) - 1, x)


# Uso:
lista = [10, 20, 30, 40, 50, 60, 70, 80]
valor = 30

# El usuario llama a la lanzadera sin preocuparse de los índices
posicion = busqueda_binaria(lista, valor)

print(f"El elemento {valor} está en la posición: {posicion}")
```

## 2. Métodos lanzadera en el contexto de la POO

En la Programación Orientada a Objetos, el uso de métodos lanzadera aplica además el principio de **ocultación de la información** (encapsulación):

1. **El método recursivo es privado** (`__metodo`): 
    - No debe ser accesible desde fuera de la clase, ya que requiere parámetros internos y su uso incorrecto podría romper la lógica del objeto.
    - No se deben pasar como parámetros los objetos (información) que forman parte del estado interno (atributos) de la instancia. 

2. **El método lanzadera es público**: 
    - Define la interfaz oficial pública de la clase para realizar la operación.

### Ejemplo: Gestión de un Padrón Municipal

Vamos a diseñar un ejemplo con una clase `Persona` y una clase `PadronMunicipal` que gestiona una lista de personas ordenadas por su número de DNI.

#### La clase `Persona`

```python
class Persona:
    def __init__(self, dni, nombre="Desconocido"):
        self.dni = dni
        self.nombre = nombre
    
    # Igualdad definida por DNI
    def __eq__(self, other):
        if not isinstance(other, Persona):
            return False
        return self.dni == other.dni
    
    # Comparaciones basadas en el DNI (para poder ordenar y buscar)
    def __lt__(self, other):
        return self.dni < other.dni
    
    def __le__(self, other):
        return self.dni <= other.dni
    
    def __gt__(self, other):
        return self.dni > other.dni
    
    def __ge__(self, other):
        return self.dni >= other.dni
    
    def __str__(self):
        return f"Persona con nombre '{self.nombre}' y DNI '{self.dni}'"
```

#### La clase `PadronMunicipal` con Método Lanzadera

```python
class PadronMunicipal:
    def __init__(self, personas):
        # Almacenamos la lista ordenada por fecha de nacimiento para búsqueda binaria
        self.__habitantes = sorted(personas)
    
    # Método recursivo privado
    def __busqueda_binaria_recursiva(self, beg, end, p):
        """
        Notar que no pasamos el parámetro "seq", ya que esa información
          forma parte del estado interno (es "self.__habitantes")
        
        "p" es una instancia de Persona con el dni que buscamos (el nombre da igual)
        """
        if beg > end:
            return None # Devolvemos None en lugar de -1
        else:
            medio = (beg + end) // 2
            if self.__habitantes[medio] == p:
                return self.__habitantes[medio] # devolvemos el objeto, y no el índice "medio"
            elif self.__habitantes[medio] < p:
                return self.__busqueda_binaria_recursiva(medio + 1, end, p)
            else:
                return self.__busqueda_binaria_recursiva(beg, medio - 1, p)

    # MÉTODO LANZADERA (Público)
    def buscar(self, dni):
        p = Persona(dni) # creamos instancia de persona con el dni proporcionado y nombre desconocido
        return self.__busqueda_binaria_recursiva(0, len(self.__habitantes) - 1, p)


habitantes = [
    Persona("11111111A", "Lídia"),
    Persona("22222222B", "Pau"),
    Persona("33333333C", "Júlia")
]
padron = PadronMunicipal(habitantes)

for dni in ["22222222B", "12345678Z"]:
    persona = padron.buscar(dni)
    print(f"Buscando DNI {dni}: {persona if persona else "No encontrado"}")
```
