### Ejercicio 1

Vamos a modelar una estructura básica para juegos de cartas, practicando los métodos especiales.

1. Crea una clase `Carta`:
   - El constructor recibirá un `valor` (1 a 12) y un `palo` ("Oros", "Copas", "Espadas", "Bastos").
     + Se deberá comprobar la validez de los datos, lanzando `ValueError` en caso contrario.
     + La comprobación de `palo` será *case-insensitive*.
     + Se almacenarán como atributos privados, que se podrán consultar (solo lectura) con propiedades.
   - Sobreescribe el método especial equivalente a `__str__` para que devuelva, por ejemplo: `"5 de Oros"`.
   - Sobreescribe el método de igualdad para que dos cartas sean idénticas si tienen el mismo valor y el mismo palo.
   - Sobreescribe el método de "menor que" (`<`) para que actúe únicamente sobre el `valor` de la carta (útil para ordenarlas).
2. Crea una clase `Mazo`:
   - Internamente gestiona una lista de `cartas` (privado, sin acceso de lectura ni escritura).
   - Implementa un método `add_carta(carta)` para añadir nuevas cartas al final de la lista.
   - Implementa un método `tiene_el_dos_de_oros()`, que devuelva `True` si el mazo contiene la carta del 2 de Oros. 
   - Sobreescribe el método especial que permita ejecutar la función `len()` sobre una instancia de `Mazo`, devolviendo el número de cartas actuales.
   - Permite el acceso por indexación (`[]`) implementando el método especial adecuado.
   - Haz que el objeto sea iterable implementando el protocolo iterador (`__iter__` y `__next__`) para poder recorrer el mazo en un bucle `for`.

En el programa principal, crea varias cartas, añade validaciones booleanas comprobando igualdades y desigualdades lógicas de cartas, agrégalas a un mazo, comprueba su longitud y si tiene el 2 de Oros, e itera el mazo mostrando las cartas.


```python
class Carta:
    def __init__(self, valor, palo):
        if not(isinstance(valor, int)) or valor < 1 or valor > 12:
            raise ValueError("Valor de carta '{valor}' no válido")
        self.__valor = valor

        if palo.lower() not in ("oros", "copas", "espadas", "bastos"): # case insensitive
            raise ValueError("Palo de carta '{palo}' no válido")
        self.__palo = palo.lower().capitalize() # Nos aseguramos que, p.e. oROs -> Oros

    @property
    def valor(self):
        return self.__valor

    @property
    def palo(self):
        return self.__palo
        
    def __str__(self):
        return f"{self.__valor} de {self.__palo}"
        
    def __eq__(self, otra):
        if not(isinstance(otra, Carta)): 
            return False
        return self.__valor == otra.__valor and self.__palo == otra.__palo
        
    def __lt__(self, otra):
        return self.__valor < otra.__valor


class Mazo:
    def __init__(self):
        self.__cartas = []
        
    def add_carta(self, carta):
        self.__cartas.append(carta)

    def tiene_el_dos_de_oros(self):
        for carta in self.__cartas:
            if carta.palo == "Oros" and carta.valor == 2:
                return True
        return False
        
    def __len__(self):
        return len(self.__cartas)
        
    def __getitem__(self, indice):
        return self.__cartas[indice]
        
    def __iter__(self):
        self.__iter_i = 0
        return self
        
    def __next__(self):
        if self.__iter_i < len(self.__cartas):
            carta = self.__cartas[self.__iter_i]
            self.__iter_i += 1
            return carta
        else:
            raise StopIteration

# Prueba principal
carta1 = Carta(5, "Oros")
carta2 = Carta(12, "Copas")
carta3 = Carta(5, "Oros")

print(f"Carta 1: {carta1}")
print(f"Carta 2: {carta2}")
print(f"Carta 3: {carta3}")

print(f"\n¿Carta 1 es igual a Carta 2?: {carta1 == carta2}")
print(f"¿Carta 1 es igual a Carta 3?: {carta1 == carta3}")
print(f"¿Carta 1 es menor que Carta 2?: {carta1 < carta2}")
print(f"¿Carta 2 es menor que Carta 1?: {carta2 < carta1}")

mazo = Mazo()
mazo.add_carta(carta1)
mazo.add_carta(carta2)
mazo.add_carta(carta3)

print(f"\nLongitud del mazo: {len(mazo)} cartas")
print(f"\nPrimera carta del mazo (índice 0): {mazo[0]}")

print(f"\nTiene el dos de Oros? {mazo.tiene_el_dos_de_oros()}")

mazo.add_carta(Carta(2, "Oros"))

print(f"\nY ahora? Tiene el dos de Oros? {mazo.tiene_el_dos_de_oros()}")

print("\nIterando sobre el mazo:")
for carta in mazo:
    print(f" - {carta}")
```

### Ejercicio 2

Vamos a modelar un usuario de un servicio, controlando su edad mediante el uso de excepciones de usuario.

1. Define dos excepciones propias creando las clases `ErrorEdadInvalida` y `ErrorMenorDeEdad`.
2. Crea una clase `Usuario` cuyo constructor reciba su `nombre` y `edad`. El proceso de validación sucederá durante la instanciación del objeto:
   - Si la edad es menor que 0 o mayor que 120, el constructor lanzar la excepción `ErrorEdadInvalida`.
   - Si la edad está entre 0 y 17 (ambos incluidos), debe lanzar la excepción `ErrorMenorDeEdad`.
   - Sobreescribe el método especial de representación en cadena para mostrar la información: `"Usuario: Lídia, Edad: 25"`.
3. En un bloque de código principal:
   - Crea una lista de cuatro tuplas de dos componentes, donde la primera es un nombre y la segunda es una edad. Usa edades límite (edad negativa, menor de edad, adulta, y supercentenaria de más de 120 años). 
   - Itera sobre la lista, intentando crear una instancia de Usuario con la información de la tupla.
   - Captura las 2 excepciones propias individualmente y muestra un mensaje informando del fallo.


```python
class ErrorEdadInvalida(Exception):
    pass

class ErrorMenorDeEdad(Exception):
    pass

class Usuario:
    def __init__(self, nombre, edad):
        if edad < 0 or edad > 120:
            raise ErrorEdadInvalida(f"La edad proporcionada ({edad}) no es realista para un humano.")
        if edad < 18:
            raise ErrorMenorDeEdad(f"El usuario '{nombre}' no cumple con la edad mínima.")
            
        self.nombre = nombre
        self.edad = edad
        
    def __str__(self):
        return f"Usuario: {self.nombre}, Edad: {self.edad}"


if __name__ == "__main__":

    # Prueba principal
    casos_prueba = [
        ("Joan", 123),
        ("Lídia", 25),
        ("Pau", 16),
        ("Júlia", -5)
    ]

    for nombre, edad in casos_prueba:
        print(f"\nIntentando crear usuario para {nombre} ({edad} años)...")

        try:
            usuario = Usuario(nombre, edad)
            print(f" > {usuario} creado correctamente.")
        except ErrorEdadInvalida as e:
            print(f" > ErrorEdadInvalida: {e}")
        except ErrorMenorDeEdad as e:
            print(f" > ErrorMenorDeEdad: {e}")
```

### Ejercicio 3

En este ejercicio vamos a modelar los objetos de un videojuego. 

Por un lado tenemos los ítems, caracterizados por su nombre, lo que pesan (en kg) y la cantidad de usos restantes (su durabilidad numérica). Estas propiedades no podrán ser cambiadas una vez el ítem es creado, pero se podrán consultar libremente. Cuando se intente imprimir por pantalla uno de estos objetos, debe mostrar visualmente su nombre junto a su peso y los usos restantes en un formato legible. El jugador puede forjar nuevos objetos mediante fusión, usando el operador `+`. El resultado es un objeto completamente nuevo, que tendrá como nombre una combinación de los dos nombres originales, además de sumar sus pesos y sus usos restantes totales. Cada objeto debe poder ser usado. Cada vez que se use, pierde un punto de usos restantes por desgaste. Llegados al punto en el que el objeto se queda sin usos restantes, la operación de uso del mismo debe abortarse de inmediato, lanzando una excepción diseñada específicamente para alertar de esta situación. 

Por otro lado, debe existir una clase que represente una mochila de objetos. Esta se inicializará con un límite máximo de capacidad de carga predefinido, que podrá ser consultado, pero no modificado. La mochila debe permitir la introducción de nuevos objetos. Cuando se intenta guardar un item, se debe comprobar si este objeto cabe o no. En caso contrario, debe abortarse la operación, lanzando una excepción diseñada específicamente para alertar de esta situación, que además, deberá almacenar el ítem que no ha podido ser almacenado. Se podrá consultar el número de ítems almacenados en la mochila con `len()`. Las mochilas deben ser iterables, para que se puedan examinar fácilmente los items almacenados en la misma.

Se deberá verificar la integridad de los datos proporcionados para crear los objetos (p.e. comprobar valores negativos).

En un bloque principal de prueba, demuestra el correcto funcionamiento construyendo e instanciando al menos dos objetos diferentes. Genera un objeto adicional como resultado de fusionarlos. Añade dichos objetos a una lista, y recórrela para ir insertándolos en una mochila, hasta que se exceda la capacidad máxima, capturando la excepción correspondiente y extrayendo de ella el objeto causante del error. A continuación, explora el contenido de la mochila, item a item, y, por último, agota la vida útil de uno de los objetos, evitando que el programa falle.


```python
class ErrorItemRoto(Exception):
    pass


class ErrorMochilaLlena(Exception):
    def __init__(self, item, mensaje=""):
        super().__init__(mensaje)
        self.item = item


class Item:
    def __init__(self, nombre, peso, usos_restantes):
        self.__nombre = nombre
        if peso < 0:
            raise ValueError("El peso no puede ser negativo")
        self.__peso = peso
        if usos_restantes < 0:
            raise ValueError("El número de usos restantes no puede ser negativo")
        self.__usos_restantes = usos_restantes
        
    @property
    def nombre(self):
        return self.__nombre
        
    @property
    def peso(self):
        return self.__peso
        
    @property
    def usos_restantes(self):
        return self.__usos_restantes
        
    def __str__(self):
        return f"[{self.__nombre}] -> Peso: {self.__peso}kg | Usos restantes: {self.__usos_restantes}"
        
    def __add__(self, otro):
        if not(isinstance(otro, Item)):
            raise ValueError("No se pueden fusionar")
        nuevo_alias = f"{self.__nombre} fusionado con {otro.__nombre}"
        nuevo_peso = self.__peso + otro.__peso
        nuevos_usos = self.__usos_restantes + otro.__usos_restantes
        return Item(nuevo_alias, nuevo_peso, nuevos_usos)
        
    def usar(self):
        if self.__usos_restantes <= 0:
            raise ErrorItemRoto("Item roto.")
        self.__usos_restantes -= 1


class Mochila:
    def __init__(self, capacidad_maxima):
        if capacidad_maxima < 0:
            raise ValueError("La capacidad máxima no puede ser negativa")
        self.__capacidad_maxima = capacidad_maxima
        self.__items = [] 
        
    @property
    def capacidad_maxima(self):
        return self.__capacidad_maxima
        
    def add_item(self, item):
        peso_total = sum(item.peso for item in self.__items)
        if peso_total + item.peso > self.capacidad_maxima:
            raise ErrorMochilaLlena(item, "No cabe!")
        self.__items.append(item)
        
    def __len__(self):
        return len(self.__items)
        
    def __iter__(self):
        self._iterador_idx = 0
        return self
        
    def __next__(self):
        if self._iterador_idx < len(self.__items):
            item = self.__items[self._iterador_idx]
            self._iterador_idx += 1
            return item
        raise StopIteration


espada = Item("Espada Desgastada", 4.0, 1)
machete = Item("Machete Enano", 1.5, 5)
fusion = espada + machete 

lista_items = [espada, machete, fusion]

mochila = Mochila(6)

try:
    for item in lista_items:
         print(f"Añadiendo ||| {item} ||| a la mochila...")
         mochila.add_item(item)
except ErrorMochilaLlena as e:
    print(f" > Error! Este item no cabe: {e.item}")

print(f"\nNúmero de elementos en la mochila: {len(mochila)}")

print("\nIterando sobre los elementos de la mochila")
for item in mochila:
    print(" >>>", item)

print("\nProvocando rotura de la espada...")
try:
    print(espada)
    espada.usar()
    print(espada)
    espada.usar()
except ErrorItemRoto as e:
    print(f"Error! {e}")
```
