# T4.3a Análisis de algoritmos recursivos

## 1. Procedimiento de análisis

El proceso de análisis del coste temporal de un algoritmo recursivo requiere seguir estos pasos:

1. Identificar la **talla** $n$ del problema.
2. Determinar la **unidad de medida de coste temporal**.
    * Debido a la dicotomía caso base/general de la recursión, para ciertas tallas $n$
      la instrucción crítica podría ejecutarse 0 veces.
    * Por ello, a fin de evitar funciones de coste temporal nulo, se recomienda usar la **unidad de pasos por programa (p.p.)**.
3. Identificar si existen **instancias** diferentes o no, y en tal caso, describir los dos más significativas (caso mejor y peor).
4. **Caracterizar la función $T(n)$** para el caso único, o para el mejor y para el peor caso.
    * Expresión matemática que permite contar el número de pasos de programa que se ejecutarán en función de $n$.
    * Se elabora mediante **relaciones de recurrencia**
    * La recurrencia se resuelve mediante eñ **método de sustitución**.
5. Proporcionar **cotas de coste temporal en notación asintótica**.



## 2. Relaciones de recurrencia y su resolución

### 2.1 Expresión del coste temporal mediante relación de recurrencia

La forma más natural de expresar el coste de un algoritmo recursivo es mediante una **relación de recurrencia**. 

Supongamos que el caso general hace una sola llamada recursiva:

$$T(n) = \left\{
\begin{array}{cl}
T_B &  \text{si} & n \; \text{ en el caso base} \\
T_G + T(n_r) &  \text{si} & n \; \text{ en el caso general} 
\end{array} \right. $$

Donde:
- $T_B$ es el coste temporal (en pasos de programa) del código (cuerpo) caso base.
- $T_G$ es el coste temporal (en pasos de programa) del código (cuerpo) del caso general.
- $n_r$ es la talla (reducida) del sub-problema que se resuelve recursivamente.
- $T(n_r)$ es el coste temporal del sub-problema, lo que define la relación de recurrencia del coste temporal.


#### Ejemplo: factorial

Ya conocemos el algoritmo factorial recursivo:

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
```

Sabemos que la talla es $n$ y que no presenta instancias diferentes.

* **Condición de talla para el caso base:** $n = 0$
* **Coste del caso base $T_B$:** 1 paso de programa (p.p.)
* **Condición de talla para el caso general:** $n > 0$
* **Coste del caso general $T_G$:** 1 p.p.
* **Talla reducida $n_r$:** tras cada llamada recursiva, la talla se reduce en una unidad ($n_r = n -1$)

Por tanto:

$$T(n) = \left\{
\begin{array}{cl}
1 \; \text{p.p.} &  \text{si} & n = 0 \\
1 + T(n - 1) \; \text{p.p.} &  \text{si} & n > 0 
\end{array} \right. $$


### 2.2 Resolución de la recurrencia mediante el método de sustitución

Una vez determinada la ecuación recursiva de coste $T(n$), se resuelve la recurrencia mediante el método de sustitución:

1. Empezar con la expresión de $T(n)$ para una talla $n$ en el caso general.
2. Obtener el coste temporal provisional tras realizar la siguiente
llamada recursiva: sustituir $T(n_r)$ y simplificar la expresión resultante.
3. Repetir el paso 2 varias veces hasta detectar un patrón.
4. Parametrizar la expresión de coste en función del número de llamadas ($i$).
5. Determinar qué valor debe tomar el parámetro $i$ para alcanzar el caso base de $T(n$).
6. Sustituir $i$ por dicho valor, y simplificar la expresión resultante.

#### Ejemplo: factorial (cont.)

1. Empezar con la expresión de $T(n)$ para una talla $n$ en el caso general.

$$T(n) = 1 + T(n-1) \; \text{p.p.}$$

2. Obtener el coste temporal provisional tras realizar la siguiente
llamada recursiva: sustituir $T(n_r)$ y simplificar la expresión resultante.

$$ T(n) = 1 + T(n-1) = 1 + (1 + T(n-2)) = 2 + T(n-2) \; \text{p.p.}$$

3. Repetir el paso 2 varias veces hasta detectar un patrón.

$$ T(n) = 1 + T(n-1) = 2 + T(n-2) = 3 + T(n-3) \; \text{p.p.}$$

4. Parametrizar la expresión de coste en función del número de llamadas ($i$).

$$T(n) = i + T(n-i) \; \text{p.p.}$$

5. Determinar qué valor debe tomar el parámetro $i$ para alcanzar el caso base de $T(n$).

$$T(n-i) = 0 \;\to\; n-i = 0; \; i = n$$

6. Sustituir $i$ por dicho valor, y simplificar la expresión resultante.

$$T(n) = i + T(n-i) = n + T(n-n) = n + T(0) = n + 1 \; \text{p.p.}$$

## 3. Análisis de algoritmos recursivos

### 3.1 Impresión de triángulos de asteriscos 

Vamos a analizar el coste de imprimir $n$ líneas de longitud creciente $1, 2, \dots n$.

```python
def print_triangulo_asteriscos(n):
    """
    Condición: n >= 1
    """
    if n == 1:
        print("*")
    else:
        for i in range(1, n+1):
            print('*', end='')
        print()
        print_triangulo_asteriscos(n-1)

print_triangulo_asteriscos(4)
```

**Análisis de complejidad temporal:**

1. **Talla**: $n$ (número de líneas).
2. **Unidad de medida**: pasos de programa.
3. **Instancias diferentes**: no.
4. **Función de coste temporal**: 
    * El coste del caso base ($n = 1$) es una instrucción `print`, cuyo coste es independiente de $n$: 1 p.p.
    * El coste del caso general ($n > 1$) es:
        * un bucle for cuyo coste es **dependiente** (OJO!) de la talla $n$:
             * Realizará $n$ pasadas: coste de $n$ p.p.
        * más una instrucción `print`: 1 p.p.
        * en total: $n + 1$ p.p.
        * nota: para simplifcar la relación de recurrencia, despreciamos ese $+1$ (para tallas grandes es irrelevante).
    * Relación de recurrencia:
        $$T(n) = \left\{
        \begin{array}{cl}
            1 \; \text{p.p.} &  \text{si} & n = 0 \\
            n + T(n - 1) \; \text{p.p.} &  \text{si} & n > 0 
        \end{array} \right. $$
    * Resolución de la recurrencia por método de sustitución:
        $$ \begin{align*}
           T(n) &= n + T(n-1) \\ 
                &= n + (n-1) + T(n-2) \\
                &= n + (n-1) + (n-2) + T(n-2) \\
                &= n + (n-1) + (n-2) + (n-3) + T(n-3) \\
                &= \dots = n + (n-1) + \cdots + (n-i+1) + T(n-i) \\ 
            \end{align*}$$
      Calculamos el valor que debe tomar $i$ para alcanzar el caso base:

      $$ T(n-i) = T(1) \; \to \; n - i = 1; i = n - 1$$

      Sustituimos y solucionamos, calculando la suma de la progresión aritmética que hemos destapado:

      $$ \begin{align*}
           T(n) &= n + (n-1) + \cdots + (n-i+1) + T(n-i) \\ 
                &= n + (n-1) + \cdots + 2 + T(1) \\
                &= n + (n-1) + \cdots + 2 + 1 \\
                &= \frac{(n+1)n}{2} = \frac{n^2 + n}{2} \; \text{p.p.}
            \end{align*}$$

5. **Cotas de complejidad**:
    * $T(n) \in \Theta(n^2)$ (coste cuadrático)


### 3.2 Búsqueda secuencial recursiva

```python
def buscar_recursivo(lista, x, ini):
    """Condición: ini >= 0"""
    if ini >= len(lista):
        return -1
    else: 
        if lista[ini] == x: 
            return ini
        else: 
            return buscar_recursivo(lista, x, ini + 1)

buscar_recursivo([1,5,7,9], 7, 0)
```

**Análisis de complejidad temporal:**

1. **Talla**: $n = len(lista) - ini$ (número de elementos por examinar).
2. **Unidad de medida**: pasos de programa.
3. **Instancias diferentes**: sí.
    * **Caso mejor**: El elemento buscado está en la posición `ini` (primera posición examinada).
    * **Caso peor**: El elemento no está en la lista.
4. **Función de coste temporal**:

    **Caso mejor:**
    * Se encuentra el elemento en la primera comparación, sin llamada recursiva.
    * $T^m(n) = 1 \;$ p.p.

    **Caso peor:**
    * El coste del caso base ($ini \geq len(lista)$, es decir, $n = 0$): 1 p.p.
    * El coste del caso general ($n > 0$): se realiza la comparación y la llamada recursiva, reduciendo la talla en 1: 1 p.p.
    * Relación de recurrencia:
        $$T^p(n) = \left\{
        \begin{array}{cl}
            1 \; \text{p.p.} &  \text{si} & n = 0 \\
            1 + T^p(n - 1) \; \text{p.p.} &  \text{si} & n > 0 
        \end{array} \right. $$
    * Resolución de la recurrencia por método de sustitución:
        $$ \begin{align*}
           T^p(n) &= 1 + T^p(n-1) \\ 
                &= 1 + 1 + T^p(n-2) = 2 + T^p(n-2)\\
                &= 3 + T^p(n-3) \\
                &= \dots = i + T^p(n-i) \\ 
            \end{align*}$$
      Calculamos el valor que debe tomar $i$ para alcanzar el caso base:

      $$ T^p(n-i) = T^p(0) \; \to \; n - i = 0; \; i = n$$

      Sustituimos:

      $$ T^p(n) = i + T^p(n-i) = n + T^p(0) = n + 1 \; \text{p.p.}$$

5. **Cotas de complejidad**:
    * $T^m(n) \in \Theta(1)$
    * $T^p(n) \in \Theta(n)$
    * $T(n) \in \Omega(1) \cap O(n)$

### 3.3 Búsqueda binaria recursiva

Igual que en el caso iterativo, la talla se divide por 2 en cada paso (reducción geométrica).

```python
def binaria_recursiva(a, x, ini, fin):
    if ini > fin:
        return -1
    medio = (ini + fin) // 2
    if a[medio] == x:
        return medio
    elif a[medio] < x:
        return binaria_recursiva(a, x, medio + 1, fin)
    else:
        return binaria_recursiva(a, x, ini, medio - 1)
```

**Análisis de complejidad temporal:**

1. **Talla**: $n = fin - ini + 1$ (número de elementos en la zona de búsqueda).
2. **Unidad de medida**: pasos de programa.
3. **Instancias diferentes**: sí.
    * **Caso mejor**: El elemento buscado está en la posición central.
    * **Caso peor**: El elemento no está en la lista.
4. **Función de coste temporal**:

    **Caso mejor:**
    * Se encuentra el elemento en la primera comparación, sin llamada recursiva.
    * $T^m(n) = 1 \;$ p.p.

    **Caso peor:**
    * El coste del caso base ($ini > fin$, es decir, $n = 0$): 1 p.p.
    * El coste del caso general ($n > 0$): se realiza el cálculo del nuevo medio, la comprobación fallida de encontrar la solución, la comparación con el elemento medio, y una llamada recursiva sobre la mitad de la zona de búsqueda, con talla $n_r = n/2$: 1 p.p.
    * Relación de recurrencia:
        $$T^p(n) = \left\{
        \begin{array}{cl}
            1 \; \text{p.p.} &  \text{si} & n = 0 \\
            1 + T^p(n / 2) \; \text{p.p.} &  \text{si} & n > 0 
        \end{array} \right. $$
    * Resolución de la recurrencia por método de sustitución:
        $$ \begin{align*}
           T^p(n) &= 1 + T^p(n/2) \\ 
                &= 1 + 1 + T^p(n/4) = 2 + T^p(n/2^2)\\
                &= 3 + T^p(n/2^3)\\
                &= \dots = i + T^p(n/2^i) \\ 
            \end{align*}$$
      Calculamos el valor que debe tomar $i$ para alcanzar el caso base:

      $$ T^p(n/2^i) = T^p(0) \; \to \; n/2^i = 0$$

      En la práctica, el caso base se alcanza cuando la zona de búsqueda se reduce a 1 elemento, y a continuación, a cero elementos, es decir cuando $n/2^i = 1$:

      $$ n/2^i = 1; \; 2^i = n; \; i = \log_2 n$$

      Y finalmente, un paso adicional para alcanzar talla cero:

      $$ \log_2 n +1 $$

      Sustituimos:

      $$ T^p(n) = i + T^p(n/2^i) = (\log_2 n + 1) + T^p(0)  = \log_2 n + 3 \; \text{p.p.}$$

5. **Cotas de complejidad**:
    * $T^m(n) \in \Theta(1)$
    * $T^p(n) \in \Theta(\log n)$
    * $T(n) \in \Omega(1) \cap O(\log n)$

### 3.4 El mínimo de una lista


#### Versión 1: Recursión lineal
Buscamos el mínimo reduciendo la talla de uno en uno.

```python
def minimo_lineal(a, ini):
    if ini == len(a) - 1:
        return a[ini]
    else:
        min_resto = minimo_lineal(a, ini + 1)
        return a[ini] if a[ini] < min_resto else min_resto
```

**Análisis de complejidad temporal:**

1. **Talla**: $n = len(a) - ini$ (número de elementos por examinar).
2. **Unidad de medida**: pasos de programa.
3. **Instancias diferentes**: no.
4. **Función de coste temporal**:
    * El coste del caso base ($ini = len(a) - 1$, es decir, $n = 1$): 1 p.p.
    * El coste del caso general ($n > 1$): se realiza la llamada recursiva y una comparación: 1 p.p. La talla se reduce en 1 ($n_r = n - 1$).
    * Relación de recurrencia:
        $$T(n) = \left\{
        \begin{array}{cl}
            1 \; \text{p.p.} &  \text{si} & n = 1 \\
            1 + T(n - 1) \; \text{p.p.} &  \text{si} & n > 1 
        \end{array} \right. $$
    * Resolución de la recurrencia por método de sustitución:
        $$ \begin{align*}
           T(n) &= 1 + T(n-1) \\ 
                &= 2 + T(n-2)\\
                &= 3 + T(n-3) \\
                &= \dots = i + T(n-i) \\ 
            \end{align*}$$
      Calculamos el valor que debe tomar $i$ para alcanzar el caso base:

      $$ T(n-i) = T(1) \; \to \; n - i = 1; \; i = n - 1$$

      Sustituimos:

      $$ T(n) = i + T(n-i) = (n-1) + T(1) = (n-1) + 1 = n \; \text{p.p.}$$

5. **Cotas de complejidad**:
    * $T(n) \in \Theta(n)$ (coste lineal)

#### Versión 2: Recursión múltiple (Divide y Vencerás)
Dividimos la lista en dos mitades y buscamos el mínimo en cada una.

```python
def minimo_multiple(a, ini, fin):
    if ini == fin:
        return a[ini]
    else:
        medio = (ini + fin) // 2
        min_izq = minimo_multiple(a, ini, medio)
        min_der = minimo_multiple(a, medio + 1, fin)
        return min_izq if min_izq < min_der else min_der
```

**Análisis de complejidad temporal:**

1. **Talla**: $n = fin - ini + 1$ (número de elementos en el sub-array).
2. **Unidad de medida**: pasos de programa.
3. **Instancias diferentes**: no.
4. **Función de coste temporal**:
    * El coste del caso base ($ini = fin$, es decir, $n = 1$): 1 p.p.
    * El coste del caso general ($n > 1$): se realizan **dos** llamadas recursivas, cada una sobre la mitad de la lista (talla $n/2$), más una comparación: 1 p.p.
    * Relación de recurrencia:
        $$T(n) = \left\{
        \begin{array}{cl}
            1 \; \text{p.p.} &  \text{si} & n = 1 \\
            1 + 2 \cdot T(n / 2) \; \text{p.p.} &  \text{si} & n > 1 
        \end{array} \right. $$
    * Resolución de la recurrencia por método de sustitución:
        $$ \begin{align*}
           T(n) &= 1 + 2 \cdot T(n/2) \\ 
                &= 1 + 2 \cdot (1 + 2 \cdot T(n/4)) = 1 + 2 + 4 \cdot T(n/4) \\
                &= 1 + 2 + 4 \cdot (1 + 2 \cdot T(n/8)) = 1 + 2 + 4 + 8 \cdot T(n/8) \\
                &= \dots = 1 + 2 + 4 + \cdots + 2^{i-1} + 2^i \cdot T(n/2^i) \\
                &= \sum_{k=0}^{i-1} 2^k + 2^i \cdot T(n/2^i) \\
                &= (2^i - 1) + 2^i \cdot T(n/2^i) 
            \end{align*}$$
      Calculamos el valor que debe tomar $i$ para alcanzar el caso base:

      $$ T(n/2^i) = T(1) \; \to \; n/2^i = 1; \; 2^i = n; \; i = \log_2 n$$

      Sustituimos:

      $$ \begin{align*} T(n) &= (2^i - 1) + 2^i \cdot T(1) \\ &= (n - 1) + n \cdot 1 = 2n - 1 \; \text{p.p.} \end{align*} $$

5. **Cotas de complejidad**:
    * $T(n) \in \Theta(n)$ (coste lineal)
