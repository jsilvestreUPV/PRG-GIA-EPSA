## 7. Funciones a trozos (recurrencias)

Para escribir una funciĂ³n definida a trozos (como una recurrencia), se usa `\left\{` con un bloque `\begin{array}...\end{array}`.

CĂ³digo:

Resultado:

$$T(n) = \left\{

     1 \;            \text{p.p.}  &  \text{si}  &  n = 0 \\

\right.$$

> **Desglose de la sintaxis**:
> - `\left\{` abre la llave grande. `\right.` la cierra de forma invisible.
> - `\begin{array}{cl}...\end{array}` crea una tabla con columnas: `c` (centrada) y `l` (alineada a la izquierda).
> - `&` separa las columnas.
> - `\\` separa las filas.
