## Ejercicio 1: Suma de menores

La siguiente función recursiva calcula la suma de los elementos de una lista `a` de enteros que son menores que un entero `x` dado.

Analiza de forma teórica el coste temporal de este algoritmo. 

```python
def suma_menores(a, pos, x):
    if pos < len(a):
        if a[pos] >= x:
            return suma_menores(a, pos + 1, x)
        else:
            return a[pos] + suma_menores(a, pos + 1, x)
    else:
        return 0

print(suma_menores([3, 1, 4, 1, 5, 9], 0, 4))  
print(suma_menores([10, 20, 30], 0, 5))          
print(suma_menores([1, 2, 3], 0, 10))             
```

### Análisis de complejidad temporal

#### Paso 1: Talla

$n = len(a) - pos$, el número de elementos pendientes de examinar.

#### Paso 2: Unidad de medida

Pasos de programa (p.p.).

#### Paso 3: Instancias significativas

No. Se debe recorrer siempre la lista entera para calcular el resultado.

#### Paso 4: Función de coste temporal

- Coste del caso base ($pos \geq len(a)$, es decir, $n = 0$): 1 p.p.
- Coste del caso general ($n > 0$): se realiza una comparación y una llamada recursiva, reduciendo la talla en 1: 1 p.p.

Relación de recurrencia:

$$T(n) = \left\{
\begin{array}{cl}
1 \; \text{p.p.} &  \text{si} & n = 0 \\
1 + T(n - 1) \; \text{p.p.} &  \text{si} & n > 0
\end{array} \right.$$

Resolución por método de sustitución:

$$\begin{align*}
T(n) &= 1 + T(n-1) \\
&= 2 + T(n-2) \\
&= 3 + T(n-3) \\
&= \dots = i + T(n-i)
\end{align*}$$

Valor de $i$ para alcanzar el caso base:

$$T(n-i) = T(0) \; \to \; n - i = 0; \; i = n$$

Sustituyendo:

$$T(n) = i + T(n-i) = n + T(0) = n + 1 \; \text{p.p.}$$

#### Paso 5: Cotas de complejidad

- $T(n) \in \Theta(n)$ (coste lineal)

## Ejercicio 2: Suma de elementos precedentes

La siguiente función recursiva `sum_prev_elems`, dado una lista `a` de enteros y una posición `pos` válida, procesa la lista en orden descendente comenzando desde `pos`, almacenando en cada posición de `a` la suma del elemento ubicado en `pos` con la suma de todos los elementos ubicados en las posiciones precedentes (desde 0 hasta `pos-1`).

Analiza de forma teórica el coste temporal de este algoritmo. 

```python
def suma(a, begin, end):
    """Función auxiliar iterativa que suma los elementos de a[begin..end]."""
    s = 0
    for i in range(begin, end + 1):
        s += a[i]
    return s

def sum_prev_elems(a, pos):
    if pos > 0:
        a[pos] = suma(a, 0, pos)
        sum_prev_elems(a, pos - 1)

a = [1, 2, 3, 4, 5]
print("Antes:", a)
sum_prev_elems(a, len(a) - 1)
print("Después:", a)
```

### Análisis de complejidad temporal

#### Paso 1: Talla

$n = pos + 1$, el número de elementos a procesar (desde la posición 0 hasta `pos`).

#### Paso 2: Unidad de medida

Pasos de programa (p.p.).

#### Paso 3: Instancias significativas

No. Se debe recorrer siempre toda la parte de la lista desde `pos` hasta 0.

#### Paso 4: Función de coste temporal

- Coste del caso base ($pos = 0$, es decir, $n = 1$): 1 p.p.
- Coste del caso general ($n > 1$): se llama a `suma(a, 0, pos)` que tiene un bucle con coste $n$ p.p. (recorre `pos + 1` elementos), y después se realiza la llamada recursiva reduciendo `pos` en 1 ($n_r = n - 1$). Coste total del caso general: $n$ p.p.

Relación de recurrencia:

$$T(n) = \left\{
\begin{array}{cl}
1 \; \text{p.p.} &  \text{si} & n = 1 \\
n + T(n - 1) \; \text{p.p.} &  \text{si} & n > 1
\end{array} \right.$$

Resolución por método de sustitución:

$$\begin{align*}
T(n) &= n + T(n-1) \\
&= n + (n-1) + T(n-2) \\
&= n + (n-1) + (n-2) + T(n-3) \\
&= \dots = n + (n-1) + (n-2) + \cdots + (n-i+1) + T(n-i)
\end{align*}$$

Valor de $i$ para alcanzar el caso base:

$$T(n-i) = T(1) \; \to \; n - i = 1; \; i = n - 1$$

Sustituyendo:

$$\begin{align*}
T(n) &= n + (n-1) + (n-2) + \cdots + 2 + T(1) \\
&= n + (n-1) + (n-2) + \cdots + 2 + 1 \\
&= \frac{(n+1) \cdot n}{2} = \frac{n^2 + n}{2} \; \text{p.p.}
\end{align*}$$

#### Paso 5: Cotas de complejidad

- $T(n) \in \Theta(n^2)$ (coste cuadrático)

## Ejercicio 3: Impresión de mitades decrecientes

La siguiente función recursiva imprime los primeros `n` elementos de una lista y, a continuación, se llama recursivamente para imprimir solo la primera mitad.

Analiza de forma teórica el coste temporal de este algoritmo. 

```python
def imprimir_reduccion(a, n):
    if n == 1:
        print(a[0])
    else:
        for i in range(n):
            print(a[i], end=' ')
        print()
        imprimir_reduccion(a, n // 2)

imprimir_reduccion([1, 2, 3, 4, 5, 6, 7, 8], 8)
```

### Análisis de complejidad temporal

#### Paso 1: Talla

$n$, el número de elementos a imprimir en la llamada actual.

#### Paso 2: Unidad de medida

Pasos de programa (p.p.).

#### Paso 3: Instancias significativas

No. El algoritmo siempre recorre los $n$ elementos e invoca la llamada recursiva con $n/2$.

#### Paso 4: Función de coste temporal

- Coste del caso base ($n = 1$): 1 p.p.
- Coste del caso general ($n > 1$): el bucle `for` recorre $n$ elementos ($n$ p.p.), y se realiza una llamada recursiva con talla $n/2$ (**reducción geométrica**). Coste total del caso general: $n$ p.p.

Relación de recurrencia:

$$T(n) = \left\{
\begin{array}{cl}
1 \; \text{p.p.} &  \text{si} & n = 1 \\
n + T(n / 2) \; \text{p.p.} &  \text{si} & n > 1
\end{array} \right.$$

Resolución por método de sustitución:

$$\begin{align*}
T(n) &= n + T(n/2) \\
&= n + n/2 + T(n/4) = n + n/2 + T(n/2^2) \\
&= n + n/2 + n/4 + T(n/2^3) \\
&= \dots = n + n/2 + n/4 + \cdots + n/2^{i-1} + T(n/2^i) \\
&= n \cdot \sum_{k=0}^{i-1} \left(\frac{1}{2}\right)^k + T(n/2^i)
\end{align*}$$

Valor de $i$ para alcanzar el caso base:

$$T(n/2^i) = T(1) \; \to \; n/2^i = 1; \; 2^i = n; \; i = \log_2 n$$

Calculamos la **suma de la progresión geométrica** $\sum_{k=0}^{i-1} (1/2)^k$ con primer término $a_0 = 1$, razón $r = 1/2$ e $i = \log_2 n$ términos:

$$\sum_{k=0}^{i-1} \left(\frac{1}{2}\right)^k = \frac{1 - (1/2)^{i}}{1 - 1/2} = 2 \left(1 - \frac{1}{2^i}\right) = 2 \left(1 - \frac{1}{n}\right) = 2 - \frac{2}{n}$$

Sustituyendo:

$$T(n) = n \cdot \left(2 - \frac{2}{n}\right) + T(1) = 2n - 2 + 1 = 2n - 1 \; \text{p.p.}$$

#### Paso 5: Cotas de complejidad

- $T(n) \in \Theta(n)$ (coste lineal)

> Notar que, aunque en cada llamada se realiza un trabajo proporcional a la talla ($n$ p.p.), la reducción geométrica de la talla hace que la suma total sea una **progresión geométrica** convergente, resultando en un coste lineal.

## Ejercicio 4: Suma de particiones

La siguiente función recursiva calcula la suma de los elementos de una lista `a` entre las posiciones `ini` y `fin`, dividiendo el problema en dos mitades iguales.

Analiza de forma teórica el coste temporal de este algoritmo.

```python
def suma(a, ini, fin):
    if ini == fin:
        return a[ini]
    else:
        medio = (ini + fin) // 2
        return suma(a, ini, medio) + suma(a, medio + 1, fin)

a = [3, 1, 4, 1, 5, 9, 2, 6]
print(f"suma({a}) = {suma(a, 0, len(a) - 1)}")
```

### Análisis de complejidad temporal

#### Paso 1: Talla

$n = fin - ini + 1$, el número de elementos en el sub-array.

#### Paso 2: Unidad de medida

Pasos de programa (p.p.).

#### Paso 3: Instancias significativas

No. El algoritmo siempre divide el array en dos mitades y realiza ambas llamadas recursivas.

#### Paso 4: Función de coste temporal

- Coste del caso base ($ini = fin$, es decir, $n = 1$): 1 p.p.
- Coste del caso general ($n > 1$): se realizan **dos** llamadas recursivas, cada una sobre la mitad de la lista (talla $n/2$), más una suma: 1 p.p.

Relación de recurrencia:

$$T(n) = \left\{
\begin{array}{cl}
1 \; \text{p.p.} &  \text{si} & n = 1 \\
1 + 2 \cdot T(n / 2) \; \text{p.p.} &  \text{si} & n > 1
\end{array} \right.$$

Resolución por método de sustitución:

$$\begin{align*}
T(n) &= 1 + 2 \cdot T(n/2) \\
&= 1 + 2 \cdot (1 + 2 \cdot T(n/4)) = 1 + 2 + 4 \cdot T(n/4) \\
&= 1 + 2 + 4 \cdot (1 + 2 \cdot T(n/8)) = 1 + 2 + 4 + 8 \cdot T(n/8) \\
&= \dots = 1 + 2 + 4 + \cdots + 2^{i-1} + 2^i \cdot T(n/2^i) \\
&= \sum_{k=0}^{i-1} 2^k + 2^i \cdot T(n/2^i) \\
&= (2^i - 1) + 2^i \cdot T(n/2^i)
\end{align*}$$

Valor de $i$ para alcanzar el caso base:

$$T(n/2^i) = T(1) \; \to \; n/2^i = 1; \; 2^i = n; \; i = \log_2 n$$

Sustituyendo:

$$\begin{align*}
T(n) &= (2^i - 1) + 2^i \cdot T(1) \\
&= (n - 1) + n \cdot 1 = 2n - 1 \; \text{p.p.}
\end{align*}$$

#### Paso 5: Cotas de complejidad

- $T(n) \in \Theta(n)$ (coste lineal)

> Notar que este resultado es análogo al del problema del cálculo del mínimo con recursión múltiple (versión 2, visto en T4.3a): la recurrencia $T(n) = 1 + 2 \cdot T(n/2)$ siempre conduce a un coste lineal $\Theta(n)$.

## Ejercicio 5: ¿Es potencia de 2?

La siguiente función recursiva comprueba si un número entero positivo `n` es una potencia de 2.

Analiza de forma teórica el coste temporal de este algoritmo.

```python
def es_potencia_de_2(n):
    if n == 1:
        return True
    if n % 2 != 0:
        return False
    return es_potencia_de_2(n // 2)

for v in [1, 2, 3, 4, 7, 8, 15, 16, 32, 100]:
    print(f"es_potencia_de_2({v}) = {es_potencia_de_2(v)}")
```

### Análisis de complejidad temporal

#### Paso 1: Talla

$n$, el valor del argumento.

#### Paso 2: Unidad de medida

Pasos de programa (p.p.).

#### Paso 3: Instancias significativas

Sí.
- **Caso mejor**: `n` es impar. Se detecta inmediatamente que no es potencia de 2 (`n % 2 != 0`) y se retorna `False` sin llamada recursiva.
- **Caso peor**: `n` es una potencia de 2 ($n = 2^k$). En cada paso, `n` es par, por lo que siempre se realiza la llamada recursiva con `n // 2` hasta llegar a $n = 1$.

#### Paso 4: Función de coste temporal

**Caso mejor:**

Se comprueba `n % 2 != 0`, resulta cierto, y se retorna sin llamada recursiva.

$$T^m(n) = 1 \; \text{p.p.}$$

**Caso peor** ($n = 2^k$):

- Coste del caso base ($n = 1$): 1 p.p.
- Coste del caso general ($n > 1$, $n$ par): se comprueba la paridad y se realiza una llamada recursiva con talla $n/2$ (**reducción geométrica**): 1 p.p.

Relación de recurrencia:

$$T^p(n) = \left\{
\begin{array}{cl}
1 \; \text{p.p.} &  \text{si} & n = 1 \\
1 + T^p(n / 2) \; \text{p.p.} &  \text{si} & n > 1
\end{array} \right.$$

Resolución por método de sustitución:

$$\begin{align*}
T^p(n) &= 1 + T^p(n/2) \\
&= 1 + 1 + T^p(n/4) = 2 + T^p(n/2^2) \\
&= 3 + T^p(n/2^3) \\
&= \dots = i + T^p(n/2^i)
\end{align*}$$

Valor de $i$ para alcanzar el caso base:

$$T^p(n/2^i) = T^p(1) \; \to \; n/2^i = 1; \; 2^i = n; \; i = \log_2 n$$

Sustituyendo:

$$T^p(n) = i + T^p(1) = \log_2 n + 1 \; \text{p.p.}$$

#### Paso 5: Cotas de complejidad

- $T^m(n) \in \Theta(1)$
- $T^p(n) \in \Theta(\log n)$ (coste logarítmico)
- $T(n) \in \Omega(1) \cap O(\log n)$
