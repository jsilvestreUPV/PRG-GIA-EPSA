# T4.4 Análisis de algoritmos de ordenación

El problema de ordenar una secuencia o conjunto de elementos es uno de los más estudiados en la historia de las ciencias de la computación. 

En este cuaderno abordaremos la explicación conceptual, implementación y el análisis del coste temporal de tres de ellos (quizás los más clásicos y populares): 

- **Selección Directa (Selection Sort)**, 
- **Inserción Directa (Insertion Sort)**, y 
- **Ordenación por Mezcla (Merge Sort)**.

Estos algoritmos implementan tres estrategias diferentes que resuelven el mismo problema, cada una con un esquema de costes temporales diferentes.

Notar que existen muchos otros algoritmos (p.e. quick sort, bubble sort, heap sort, block sort, etc.). Si te interesa, [aquí tienes más información](https://en.wikipedia.org/wiki/Sorting_algorithm).


## 1. Algoritmo de selección directa (Selection Sort)

A grandes rasgos, para ordenar una secuencia `seq` de elementos de forma ascendente, la selección directa sigue la siguiente estrategia:
- Buscar el **mínimo** de `seq` y llevarlo a la posición 0, intercambiándolo con el elemento actual en esa posición.
- Buscar el **segundo mínimo** de `seq` y llevarlo a la posición 1, intercambiándolo.
- Buscar el **tercer mínimo** de `seq` y llevarlo a la posición 2.
- ... y así sucesivamente, hasta que toda la secuencia quede ordenada.

En cada pasada `i`, la región `seq[0..i-1]` tiene los elementos más pequeños de la secuencia y se encuentra completamente ordenada.

  <a href="https://youtu.be/92BfuxHn2XE" target="_blank">
    <img src="https://img.youtube.com/vi/92BfuxHn2XE/0.jpg" alt="Video Selection Sort">
  </a><br>
  <i>Vídeo con traza visual del algoritmo de Selección Directa</i>
</div>



```python
def selection_sort(seq):
    for i in range(len(seq) - 1):
        posMin = i
        for j in range(i + 1, len(seq)):
            if seq[j] < seq[posMin]:
                posMin = j
        aux = seq[posMin]
        seq[posMin] = seq[i]
        seq[i] = aux

seq = [14, 23, 1, 10, 2]
print("Antes de ordenar:", seq)
selection_sort(seq)
print("Tras selection sort:", seq)
```

### 1.1 Análisis del coste temporal: Selection Sort

#### 1. Talla del problema

$n$: número de elementos de la secuencia `len(seq)`.

#### 2. Unidad de medida

Usaremos como instrucción crítica la condición `seq[j] < seq[posMin]`.

#### 3. Instancias significativas

No hay diferentes instancias. Independientemente de si la secuencia está ordenada, ordenada inversamente o desordenada, se realizan siempre $n - 1$ pasadas, y el coste de cada una sólo depende de los elementos que restan en `seq[i..n-1]`. Siempre recorremos todos los elementos de la sub-secuencia para encontrar el mínimo.

#### 4. Función de coste temporal

Calculamos el número de veces que se ejecuta la instrucción crítica sumando para cada iteración del bucle `i`, las repeticiones del bucle `j`.

$$T(n) = \sum_{i=0}^{n-2} \sum_{j=i+1}^{n-1} 1 = \sum_{i=0}^{n-2} ( (n-1) - (i+1) + 1 ) = \sum_{i=0}^{n-2} (n - 1 - i) = \frac{((n-1) + 1)\cdot n}{2} = \frac{n^2}{2} \; \text{i.c.}$$

#### 5. Cotas de complejidad

El coste asintótico del algoritmo *Selection Sort* es:
$$T(n) \in O(n^2)$$



## 2. Algoritmo de inserción directa (Insertion Sort)

Para ordenar una secuencia `seq`, este método expande progresivamente hacia la derecha la zona ordenada `seq[0..i - 1]`. 

A diferencia de la *selección directa*, no garantiza que la zona ordenada contenga los mínimos de la secuencia, sino simplemente que los elementos en esa región están ordenadas entre sí.

En cada pasada, se toma el elemento `seq[i]` y se inserta **ordenadamente** en la posición correspondiente dentro de los elementos que le preceden en `seq[0..i-1]`, desplazando hacia la derecha aquellos que sean mayores.

<div align="center">
  <a href="https://youtu.be/8oJS1BMKE64" target="_blank">
    <img src="https://img.youtube.com/vi/8oJS1BMKE64/0.jpg" alt="Video Insertion Sort">
  </a><br>
  <i>Vídeo con traza visual del algoritmo de Inserción Directa</i>
</div>



```python
def insertion_sort(seq):
    for i in range(1, len(seq)):
        aux = seq[i]
        j = i - 1
        while j >= 0 and seq[j] > aux:
            seq[j + 1] = seq[j]
            j -= 1
        seq[j + 1] = aux

seq = [14, 23, 1, 10, 2]
print("Antes de ordenar:", seq)
insertion_sort(seq)
print("Tras insertion sort:", seq)
```

### 2.1 Análisis del coste temporal: Insertion Sort

#### 1. Talla del problema
$n$: número de elementos de la secuencia `len(seq)`.

#### 2. Unidad de medida

Usaremos como instrucción crítica la guarda del bucle interno: `j >= 0 and seq[j] > aux`.  

#### 3. Instancias significativas

El algoritmo de Inserción Directa **SÍ** presenta diferentes instancias, ya que el bucle `while` se ejecuta un número distinto de veces según la ordenación previa de los datos:

- **Mejor caso**: El secuencia ya está inicialmente ordenada. El elemento `seq[i]` (`aux`) ya está en su posición de inserción (es mayor o igual que todos los elementos precedentes), por lo que la condición `seq[j] > aux` fallará a la primera y el bucle interior no se ejecutará ninguna vez.
- **Peor caso**: El secuencia está inicialmente ordenada en orden inverso. El elemento `seq[i]` (`aux`) deberá ser insertado siempre en la posición 0, provocando que la condición `seq[j] > aux` se evalúe `i+1` veces para llegar a insertar al principio, ejecutando todas las posibles pasadas del bucle interior.

#### 4. Función de coste temporal

- **Mejor caso:** En cada pasada de $i$ desde $1$ hasta $n-1$, la guarda del bucle `while` se evalúa sólo 1 vez y falla.

    $$T^m(n) = \sum_{i=1}^{n-1} 1 = n - 1 \; \text{i.c.}$$

- **Peor caso:** En la pasada $i$, la guarda del bucle `while` se evalúa para $j = i-1, i-2, \ldots, 0, -1$, es decir, se evalúa $i+1$ veces (incluyendo la vez que sale al fallar la condición `j >= 0`).

$$T^p(n) = \sum_{i=1}^{n-1} \left ( \sum_{j=0}^{i-1} 1 + 1 \right ) = \sum_{i=1}^{n-1} (i + 1) = \frac{(2+n)(n-1)}{2} = \frac{n^2 + n -2}{2} \; \text{i.c.}$$

#### Cotas de complejidad

- $T^m(n) \in \Theta(n)$
- $T^p(n) \in \Theta(n^2)$
- $T(n) \in \Omega(n) \cap O(n^2)$



## 3. Algoritmo de ordenación por mezcla (Merge Sort)

La ordenación por mezcla aplica la estrategia recursiva de **Divide y Vencerás**:
1. **Dividir**: Partir la secuencia original `seq` en dos mitades (dos sub-secuencias) con aproximadamente la misma cantidad de elementos.
2. **Resolver (Vencer)**: Ordenar recursivamente ambas sub-secuencias (solucionar problemas más pequeños).
3. **Combinar**: Mezclar ordenadamente los dos sub-secuencias ordenadas para formar la secuencia completa, utilizando el algoritmo de mezcla natural (visto en el cuaderno [T4.2a](./T4.2a%20Análisis%20de%20algoritmos%20iterativos.ipynb)).
    - Nota: aquí usaremos una versión adaptada del algoritmo de mezcla natural `mezcla_one_seq()` para que trabaje con una única secuencia de entrada, en lugar de dos.

<div align="center">
  <a href="https://youtu.be/ZRPoEKHXTJg" target="_blank">
    <img src="https://img.youtube.com/vi/ZRPoEKHXTJg/0.jpg" alt="Video Merge Sort">
  </a><br>
  <i>Vídeo con traza visual del algoritmo de Merge Sort</i>
</div>


```python
def mezcla_one_seq(seq, ini, medio, fin):
    temp = []
    i = ini
    j = medio + 1
    
    while i <= medio and j <= fin:
        if seq[i] <= seq[j]:
            temp.append(seq[i])
            i += 1
        else:
            temp.append(seq[j])
            j += 1
            
    while i <= medio:
        temp.append(seq[i])
        i += 1
        
    while j <= fin:
        temp.append(seq[j])
        j += 1
        
    for k in range(len(temp)):
        seq[ini + k] = temp[k]

def merge_sort(seq, ini, fin):
    if ini < fin:
        medio = (ini + fin) // 2
        merge_sort(seq, ini, medio)
        merge_sort(seq, medio + 1, fin)
        mezcla_one_seq(seq, ini, medio, fin)

seq = [14, 23, 1, 10, 2]
print("Antes de ordenar:", seq)
merge_sort(seq, 0, len(seq) - 1)
print("Tras merge sort:", seq)
```

### 3.1 Análisis del coste temporal: Merge Sort

#### 1. Talla del problema

$n$: número de elementos de la sub-secuencia a ordenar en cada llamada, que podemos expresar como `fin - ini + 1`.

#### 2. Unidad de medida

Usaremos pasos de programa.

#### 3. Instancias significativas

El coste temporal no depende de ninguna propiedad especial de los elementos ni del grado de ordenación inicial (la división es puramente posicional). Por tanto, no diferenciamos entre instancias (no hay mejor ni peor caso).

#### 4. Función de coste temporal 

Relación de recurrencia:

$$
T(n) =
\begin{cases}
1 & \text{si } n \le 1 \\
2 \cdot T(\frac{n}{2}) + n & \text{si } n > 1 
\end{cases}
$$

> Recordar que mezcla natural tiene un coste lineal con respecto a la talla, $n$ p.p.

Resolución por el método de sustitución:

$$
\begin{align*}
T(n) &= 2 \cdot T \left(\frac{n}{2}\right) + n   \\
     &= 2 \cdot \left[ 2 \cdot T(\frac{n}{4}) + \frac{n}{2} \right] + n    \\
     &= 4 \cdot T(\frac{n}{4}) + 2n \\
     &= 4 \cdot \left[ 2 \cdot T(\frac{n}{8}) + \frac{n}{4} \right] + 2n    \\
     &= 8 \cdot T(\frac{n}{8}) + 3n \\
     &= \dots \\
     &= 2^i \cdot T\left(\frac{n}{2^i}\right) + i \cdot n

\end{align*}     
$$


Valor de $i$ para alcanzar el caso base:

$$\frac{n}{2^i} = 1 \implies n = 2^i \implies i = \log_2 (n)$$

Sustituimos $i$:
$$
\begin{align*}
T(n) &= 2^i \cdot T\left(\frac{n}{2^i}\right) + i \cdot n \\
     &= 2^{\log_2(n)} \cdot T\left(\frac{n}{2^{\log_2(n)} }\right) + \log_2(n) \cdot n \\
     &= n \cdot T(1) + n \cdot \log_2(n) \\
     & = n + n\cdot \log_2(n) \; \text{p.p.}
\end{align*}
$$

#### 5. Cotas de complejidad

$$T(n) \in \Theta(n \log n)$$
