### Ejercicio 1: Distancia entre puntos
Dado dos puntos definidos por tuplas (x, y), calcula la distancia euclidiana entre ellos.

```python
import math

punto_a = (1, 2)
punto_b = (4, 6)

distancia = math.sqrt((punto_b[0] - punto_a[0])**2 + (punto_b[1] - punto_a[1])**2)
print(f"La distancia es: {distancia}")
```

### Ejercicio 2: Estadísticas de notas
Escribe una función que reciba una lista de notas y devuelva una tupla con (nota_minima, nota_maxima, media).

```python
def estadisticas_notas(notas):
    if not notas:
        return (None, None, 0.0)
    return (min(notas), max(notas), sum(notas) / len(notas))

mis_notas = [5.5, 9.0, 7.5, 8.0, 6.0]
minima, maxima, media = estadisticas_notas(mis_notas)

print(f"Mínima: {minima}, Máxima: {maxima}, Media: {media:.2f}")
```

### Ejercicio 3: Tuplas como claves
Crea un diccionario donde las claves sean tuplas que representen coordenadas (latitud, longitud) y el valor sea el nombre de una ciudad. Accede a un elemento.

```python
ciudades = {
    (39.4699, -0.3763): "Valencia",
    (40.4168, -3.7038): "Madrid",
    (41.3851, 2.1734): "Barcelona"
}

coord_valencia = (39.4699, -0.3763)
print(f"En {coord_valencia} está {ciudades[coord_valencia]}")
```

### Ejercicio 4: Ordenar lista de tuplas
Dada una lista de tuplas donde cada tupla es `(nombre, edad)`, ordena la lista por edad de menor a mayor.

```python
personas = [("Ana", 30), ("Luis", 25), ("Marta", 35), ("Pedro", 20)]

# Usamos una función lambda para indicar que ordenamos por el segundo elemento (index 1)
personas_ordenadas = sorted(personas, key=lambda x: x[1])

print("Lista ordenada por edad:")
for p in personas_ordenadas:
    print(p)
```
