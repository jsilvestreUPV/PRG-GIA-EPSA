### Ejercicio 1: Distancia entre puntos
Escribe una función que, dados dos puntos bidimensionales $\boldsymbol{x}$ e $\boldsymbol{y}$ definidos por tuplas, calcule y devuelva la distancia euclidiana entre ellos:

$$d(\boldsymbol{x}, \boldsymbol{y}) = \sqrt{(x_1 - y_1)^2 + (x_2 - y_2)^2}$$

Para ello, importa la librería [`math`](https://docs.python.org/es/3/library/math.html) y usa las funciones [`sqrt()`](https://docs.python.org/es/3/library/math.html#math.sqrt) y [`pow()`](https://docs.python.org/es/3/library/math.html#math.pow) (o alternativamente el operador incorporado `**`). 

Añade código principal para probar la función y que imprima por pantalla el resultado.

```python
import math

def distancia(x, y):
    return math.sqrt((y[0] - x[0])**2 + (y[1] - x[1])**2)

punto_a = (1, 2)
punto_b = (4, 6)

print(f"La distancia entre {punto_a} y {punto_b} es: {distancia(punto_a, punto_b)}")
```

### Ejercicio 2: Estadísticas de notas
Escribe una función que reciba una lista de notas y devuelva una tupla con (nota_minima, nota_maxima, media). Añade código principal que llame la función e imprima por pantalla el resultado.

Puedes usar las funciones incorporadas [`max()`](https://docs.python.org/es/3/library/functions.html#max), [`min()`](https://docs.python.org/es/3/library/functions.html#min), y [`sum()`](https://docs.python.org/es/3/library/functions.html#sum).

```python
def estadisticas_notas(notas):
    if not notas:
        return (None, None, 0.0)
    return (min(notas), max(notas), sum(notas) / len(notas))

mis_notas = [5.5, 9.0, 7.5, 8.0, 6.0]
minima, maxima, media = estadisticas_notas(mis_notas)

print(f"Mínima: {minima}, Máxima: {maxima}, Media: {media:.2f}")
```

### Ejercicio 3: Tuplas como claves
Crea un diccionario donde las claves sean tuplas que representen un viaje de avión (origen, destino) y el valor sea el precio del vuelo en euros. Accede a un elemento e imprime la información por pantalla.

```python
vuelos = {
    ("Alcoi", "València"): 35,
    ("València", "Barcelona"): 72,
    ("Barcelona", "Bocairent"): 42
}

clave = ("Barcelona", "Bocairent")
print(f"El vuelo {clave[0]}-{clave[1]} cuesta {vuelos[clave]} Euros.")
```

### Ejercicio 4 (opcional, pro/avanzado): Ordenar lista de tuplas
Escribe una función que, dada una lista de tuplas donde cada tupla es `(nombre, edad)`, imprime los nombres de las personas ordenadas por edad, de menor a mayor. Añade código principal que cree una lista con personas desordenadas por edad, y que llame a la función.

Para ello, haz uso de la función incorporada [`sorted()`](https://docs.python.org/es/3/library/functions.html#sorted), pasándole como parámetro nombrado `key` una [función lambda](https://docs.python.org/es/3/glossary.html#term-lambda) que identifique el campo clave que se usa para ordenar la lista.

```python
def imprime_menor_a_mayor_edad(personas):
    # Usamos una función lambda para indicar que ordenamos por el segundo elemento de la tupla (index 1)
    for nombre, edad in sorted(personas, key=lambda x: x[1]):
        print(nombre)

lista_personas = [("Lídia", 30), ("Joan", 25), ("Pau", 35), ("Júlia", 20)]
imprime_menor_a_mayor_edad(lista_personas)
```
