## Ejemplos de tipos básicos

```python
# Entero (int)
x = 10
print(f"x es {x} y su tipo es {type(x)}")
```

```python
# Flotante (float)
y = 3.14
print(f"y es {y} y su tipo es {type(y)}")
```

```python
# Booleano (bool)
b = True
print(f"b es {b} y su tipo es {type(b)}")
```

```python
# Cadena de caracteres (str)
s = "Hola Mundo"
print(f"s es '{s}' y su tipo es {type(s)}")
```

```python
# Lista (list)
lista = [1, 2, 3, "cuatro"]
print(f"lista es {lista} y su tipo es {type(lista)}")
```

```python
# Diccionario (dict)
diccionario = {"clave": "valor", "siete": 7, 3:"Marzo"}
print(f"diccionario es {diccionario} y su tipo es {type(diccionario)}")
```
