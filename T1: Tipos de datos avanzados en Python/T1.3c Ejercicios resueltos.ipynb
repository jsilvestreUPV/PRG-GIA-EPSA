### Ejercicio 1: Números pares infinitos
Crea un generador que produzca una secuencia infinita de números pares.

```python
def pares_infinitos():
    n = 0
    while True:
        yield n
        n += 2

gen_pares = pares_infinitos()
for _ in range(5):
    print(next(gen_pares))
```

### Ejercicio 2: Fibonacci
Crea un generador que devuelva los números de la secuencia de Fibonacci hasta un límite `n`.

```python
def fibonacci(limite):
    a, b = 0, 1
    while a < limite:
        yield a
        a, b = b, a + b

print(list(fibonacci(50)))
```

### Ejercicio 3: Lectura de líneas filtradas
Simula un generador que procesa una lista de frases (como si fuera un archivo) y devuelve solo aquellas que contienen una palabra clave, convirtiéndolas a mayúsculas.

```python
datos = [
    "Error en la linea 1",
    "Warning: carga alta",
    "Error: fallo de conexión",
    "Info: todo correcto"
]

def filtrar_log(lineas, palabra_clave):
    for linea in lineas:
        if palabra_clave in linea:
            yield linea.upper()

errores = filtrar_log(datos, "Error")
for e in errores:
    print(e)
```

### Ejercicio 4: Generador de rangos personalizados
Implementa tu propia versión de `range(start, stop, step)` usando un generador.

```python
def mi_range(start, stop, step=1):
    actual = start
    while actual < stop:
        yield actual
        actual += step

print(list(mi_range(1, 10, 2)))
```
