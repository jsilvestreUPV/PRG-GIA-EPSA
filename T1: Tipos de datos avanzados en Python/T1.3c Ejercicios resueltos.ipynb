### Ejercicio 1: Eliminación de duplicados
Dada una lista con nombres repetidos, obtén una lista con los nombres únicos ordenados alfabéticamente.

Para ello, haz uso de la función incorporada [`sorted()`](https://docs.python.org/es/3/library/functions.html#sorted).

```python
nombres = ["Pau", "Júlia", "Pau", "Joan", "Júlia", "Lídia"]
nombres_unicos = sorted(set(nombres))
print(nombres_unicos)
```

### Ejercicio 2: Análisis de textos

Escribe una función que reciba dos frases como parámetros y devuelva una lista con las palabras que aparecen en ambas frases. Añade código principal para probar la función.

Para ello, puedes usar el método [`split()`](https://docs.python.org/es/3/library/stdtypes.html#str.split) para dividir las frases en palabras y luego usar la aritmética de conjuntos para encontrar las palabras comunes.

Por ejemplo, si las frases son `"Hola Bocairent"` y `"Hola Alcoi"`, la función devolverá la lista `["Hola"]`.

```python
def palabras_comunes(frase1, frase2):
    palabras1 = set(frase1.split())
    palabras2 = set(frase2.split())
    comunes = palabras1 & palabras2
    return list(comunes)

frase1 = "En Alcoi se come olleta"
frase2 = "En Bocairent se come bajoques farcides"

comunes = palabras_comunes(frase1, frase2)
print(f"Palabras comunes: {comunes}")
```

### Ejercicio 3: Elementos únicos en múltiples listas
Tienes tres listas de números. Encuentra los números que están presentes en al menos una de las listas, pero excluyendo aquellos que están en las tres listas simultáneamente.

```python
l1 = [1, 2, 3, 4]
l2 = [2, 3, 4, 5]
l3 = [3, 4, 5, 6]

s1, s2, s3 = set(l1), set(l2), set(l3)

en_alguna = s1 | s2 | s3
en_todas = s1 & s2 & s3

resultado = en_alguna - en_todas
print(f"Números en alguna de las tres listas pero no en todas: {resultado}")
```
