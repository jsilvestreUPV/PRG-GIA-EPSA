### Ejercicio 1: Eliminación de duplicados
Dada una lista con nombres repetidos, obtén una lista con los nombres únicos ordenados alfabéticamente.

```python
nombres = ["Ana", "Juan", "Ana", "Pedro", "Juan", "Maria"]
nombres_unicos = sorted(list(set(nombres)))
print(nombres_unicos)
```

### Ejercicio 2: Análisis de textos
Dadas dos frases, encuentra las palabras que aparecen en ambas frases.

```python
frase1 = "el gato come pescado"
frase2 = "el perro come carne"

palabras1 = set(frase1.split())
palabras2 = set(frase2.split())

comunes = palabras1 & palabras2
print(f"Palabras comunes: {comunes}")
```

### Ejercicio 3: Control de acceso
Imagina dos grupos de usuarios: `admin` y `editores`. Verifica si un usuario "Pepe" tiene permisos (es decir, pertenece a alguno de los grupos). Luego verifica si todos los `admin` son también `editores` (subconjunto).

```python
admin = {"Ana", "Carlos"}
editores = {"Ana", "Carlos", "Beatriz", "David"}
usuario = "Pepe"

tiene_permiso = usuario in (admin | editores)
print(f"¿Pepe tiene permiso? {tiene_permiso}")

es_subconjunto = admin.issubset(editores)
print(f"¿Todos los admin son editores? {es_subconjunto}")
```

### Ejercicio 4: Elementos únicos en múltiples listas
Tienes tres listas de números. Encuentra los números que están presentes en al menos una de las listas, pero excluyendo aquellos que están en las tres listas simultáneamente.

```python
l1 = [1, 2, 3, 4]
l2 = [3, 4, 5, 6]
l3 = [4, 6, 7, 8]

s1, s2, s3 = set(l1), set(l2), set(l3)

en_alguna = s1 | s2 | s3
en_todas = s1 & s2 & s3

resultado = en_alguna - en_todas
print(f"En alguna pero no en todas: {resultado}")
```
