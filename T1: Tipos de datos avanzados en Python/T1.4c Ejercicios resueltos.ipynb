### Ejercicio 1: Números pares finitos
Crea un generador que produzca una secuencia finita de números pares.

Añade código principal que imprimar los primeros 5 números pares.

```python
def pares_finitos(n):
    for i in range(n):
        yield i * 2

for par in pares_finitos(5):
    print(par)
```

### Ejercicio 2: Números pares infinitos
Crea un generador que produzca una secuencia infinita de números pares.

A continuación, escribe una función que reciba un número entero `n` y devuelva la suma de los primeros `n` números pares.

Añade código principal para probar la función. 

```python
def pares_infinitos():
    n = 0
    while True:
        yield n
        n += 2

def suma_pares(n):
    suma = 0
    gen = pares_infinitos()
    for _ in range(n):
        suma += next(gen)
    return suma

print(suma_pares(5))
```

### Ejercicio 3: Fibonacci
Crea un generador que devuelva los números de la secuencia de Fibonacci hasta un valor límite `n`.

```python
def fibonacci(n):
    a, b = 0, 1
    while a < n:
        yield a
        a, b = b, a + b

print(list(fibonacci(50)))
```

### Ejercicio 4: Generador de rangos personalizados
Implementa tu propia versión de [`range(start, stop, step)`](https://docs.python.org/es/3/library/stdtypes.html#range) usando un generador.

Por ejemplo, `range(1, 10, 2)` debería generar los números 1, 3, 5, 7.

```python
def mi_range(start, stop, step=1):
    actual = start
    while actual < stop:
        yield actual
        actual += step

print(list(mi_range(1, 10, 2)))
```
