### Ejercicio 1: Matriz identidad modificada
Crea una matriz identidad de 5x5 (unos en la diagonal, ceros fuera) y multiplica todos sus elementos por 10.

```python
# np.eye crea la matriz identidad
identidad = np.eye(5)
resultado = identidad * 10
print(resultado)
```

### Ejercicio 2: Filtrado (Boolean Indexing)
Dado un array de números aleatorios, selecciona solo aquellos que sean mayores que 0.5.

```python
# Fijamos la semilla para reproducibilidad
np.random.seed(42)
aleatorios = np.random.rand(5)  # 5 números entre 0 y 1
print(f"Originales: {aleatorios}")

filtro = aleatorios > 0.5
seleccion = aleatorios[filtro]
print(f"Mayores que 0.5: {seleccion}")
```

### Ejercicio 3: Normalización de datos
Normaliza un vector de datos para que estén en el rango [0, 1] usando la fórmula: `(x - min) / (max - min)`.

```python
datos = np.array([10, 20, 5, 40, 30])

min_val = datos.min()
max_val = datos.max()

normalizados = (datos - min_val) / (max_val - min_val)
print(f"Datos normalizados: {normalizados}")
```

### Ejercicio 4: Estadísticas por columnas
Dada una matriz que representa (filas=alumnos, columnas=asignaturas), calcula la nota media de cada asignatura.

```python
notas = np.array([
    [5, 8, 9],
    [6, 7, 5],
    [8, 9, 8],
    [4, 5, 5]
])

# axis=0 opera colapsando las filas (es decir, calcula por columnas)
medias_asignaturas = notas.mean(axis=0)
print(f"Media Asignatura 1: {medias_asignaturas[0]}")
print(f"Media Asignatura 2: {medias_asignaturas[1]}")
print(f"Media Asignatura 3: {medias_asignaturas[2]}")
```
