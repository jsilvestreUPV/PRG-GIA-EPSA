Por convención, se importa el módulo `pyplot` con el alias `plt`.

```python
import matplotlib.pyplot as plt # típicamente se importa con el alias plt
import numpy as np
```

## 1. Gráficos básicos

La función más sencilla es `plot`, que crea un gráfico de líneas.

```python
x = np.linspace(0, 10, 100) # array de 100 valores entre 0 y 10 espaciados linealmente
y = np.sin(x) # array con el seno de todos los valores del array x

plt.plot(x, y)
plt.show()
```

## 2. Estructura de una figura

Matplotlib opera con una jerarquía de objetos. Los principales son:
- **Figure**: El contenedor de nivel superior para todos los elementos del gráfico.
- **Axes**: El área donde se dibujan los datos (lo que normalmente pensamos como un 'gráfico'). Una figura puede contener múltiples Axes.

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Gráfica de $sin(x)$") # entre $$: notación math LaTeX
ax.set_xlabel("$x$")
ax.set_ylabel("$sin(x)$")
plt.show()
```

## 3. Tipos especiales de gráficas

Además de dibujar puntos (y trazos), matplotlib también permite crear otros tipos de gráficas.

### 3.1. Gráfica de dispersión (Scatter plot)

```python
x = np.random.rand(50) # array con 50 floats aleatorios entre 0 y 1
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.show()
```

### 3.2. Gráfica de barras (Bar plot)

```python
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 56, 12]

plt.bar(categorias, valores)
plt.show()
```

### 3.3. Histograma

```python
data = np.random.randn(1000) # devuelve array con 1000 valores aleatorios siguiendo una distribución normal

plt.hist(data, bins=30, edgecolor='black')
plt.show()
```

## 4. Personalización y Subplots

Podemos combinar múltiples gráficos en una sola figura y personalizar sus estilos.

```python
x = np.linspace(0, 10, 100)

fig, axs = plt.subplots(2, 2, figsize=(10, 8)) # Figura de 2x2 gráficas

axs[0, 0].plot(x, np.sin(x), 'r')
axs[0, 0].set_title('Seno')

axs[0, 1].plot(x, np.cos(x), 'g')
axs[0, 1].set_title('Coseno')

axs[1, 0].plot(x, np.exp(x), 'b')
axs[1, 0].set_title('Exponencial')

axs[1, 1].plot(x, -x**2, 'k')
axs[1, 1].set_title('Parábola invertida')

plt.tight_layout()
plt.show()
```
