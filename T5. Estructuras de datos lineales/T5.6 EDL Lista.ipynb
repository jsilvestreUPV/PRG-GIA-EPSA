# T5.6 EDL Lista (*List*)

## 1. Concepto

Una **lista** (*list*) es una secuencia de elementos en la que el acceso y la modificación pueden realizarse en **cualquier posición**. Es la estructura de datos lineal más flexible.

A diferencia de la Pila (solo acceso al tope) y la Cola (solo acceso a los extremos), la Lista permite insertar, eliminar y consultar en cualquier posición de la secuencia.

## 2. Interfaz

| Método | Descripción | Coste (enlazada) | Coste (array) |
|--------|-------------|:-:|:-:|
| `List()` | Crea una lista vacía | $\Theta(1)$ | $\Theta(1)$ |
| `insert(pos, e)` | Inserta `e` en la posición `pos` | $\Omega(1) \cap O(n)$ | $\Omega(1) \cap O(n)$ |
| `append(e)` | Inserta `e` al final | $\Theta(n)$ (*) | $\Theta(1)$ |
| `remove(pos)` | Elimina y devuelve el dato en `pos` | $\Omega(1) \cap O(n)$ | $\Omega(1) \cap O(n)$ |
| `get(pos)` | Devuelve el dato en `pos` | $\Omega(1) \cap O(n)$ | $\Theta(1)$ |
| `search(e)` | Posición de la 1ª ocurrencia de `e` (-1 si no) | $\Omega(1) \cap O(n)$ | $\Omega(1) \cap O(n)$ |
| `empty()` | Indica si la lista está vacía | $\Theta(1)$ | $\Theta(1)$ |
| `size()` | Número de elementos | $\Theta(1)$ | $\Theta(1)$ |

(*) Se podría conseguir $\Theta(1)$ manteniendo un puntero al último nodo.

## 6. Comparativa de implementaciones

| Operación / Aspecto | Nodos enlazados | Array |
|-----------|:-:|:-:|
| `get(pos)` | $\Omega(1) \cap O(n)$ — recorrido | $\Theta(1)$ — acceso directo |
| `insert(pos, e)` | $\Omega(1) \cap O(n)$ — búsqueda posición | $\Omega(1) \cap O(n)$ — desplazamiento |
| `append(e)` | $\Theta(n)$ (*) | $\Theta(1)$ |
| `remove(pos)` | $\Omega(1) \cap O(n)$ — búsqueda posición | $\Omega(1) \cap O(n)$ — compactación |
| `search(e)` | $\Omega(1) \cap O(n)$ | $\Omega(1) \cap O(n)$ |
| **Uso de memoria** | Variable (crece/decrece dinámicamente) + sobrecoste por referencias (despreciable) | Fija (reserva `MAX` posiciones) |
| **Talla máxima** | Sin límite (salvo memoria del sistema) | Limitada a `MAX` |


(*) `append` en nodos enlazados requiere recorrer hasta el final. Se podría mejorar a $\Theta(1)$ manteniendo un puntero `__last`.

> **Observación**: La elección de la implementación depende del **patrón de uso**:
> - Si predominan los **accesos por posición** → array.
> - Si predominan las **inserciones/eliminaciones** → nodos enlazados.

Al igual como con `StackArray` y `QueueArray`, la implementación con arrays presenta el inconveniente de la estimación del tamaño máximo. 

Se podría implementar una variante de `insert()` y `append()` que permita redimensionar automáticamente el array, como vimos en la [Sección 4.1 del T5.3 sobre gestión de secuencias con memoria contigua](./T5.3%20Gestión%20de%20secuencias%20con%20memoria%20contigua.ipynb), pero entonces, el coste de ambas operaciones pasaria a ser $\Omega(1) \cap O(n)$.
