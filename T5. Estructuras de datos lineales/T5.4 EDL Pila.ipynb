# T5.4 EDL Pila (*Stack*)

## 1. Concepto

Una **pila** (*stack*) es una secuencia de elementos que se gestiona siguiendo un criterio **LIFO** (*Last In, First Out*): el último elemento en entrar es el primero en salir.

- Solo se tiene acceso al **tope** (*top*) de la pila.
- Los elementos solo se pueden **apilar** (*push*) encima del tope anterior.
- Solo se puede **desapilar** (*pop*) el elemento que está en el tope.

</div>

## 2. Interfaz

Una EDL tipo pila ofrece la siguiente interfaz de métodos:

| Método | Descripción |
|--------|-------------|
| `Stack()` | Crea una pila vacía |
| `push(e)` | Apila el elemento `e` en el tope |
| `pop()` | Desapila y devuelve el tope. Lanza excepción si vacía |
| `peek()` | Devuelve el tope sin desapilarlo. Lanza excepción si vacía |
| `empty()` | Indica si la pila está vacía |
| `size()` | Devuelve el número de elementos apilados |

Todas estas operaciones se pueden implementar con un coste temporal constante $\Theta(1)$.

## 3. Implementación con nodos enlazados: `StackLinked`

Implementamos la pila mediante una secuencia de nodos enlazados, ordenados del tope a la base. 

La inserción y extracción se realizan siempre en **cabeza** de la secuencia, que estará demarcada por un atributo `__top` que apuntará al primer nodo. 

Mantendremos un atributo `__size` con el tamaño de la pila para poder implementar la operación `size()` en tiempo constante (sin necesidad de recorrer toda la secuencia).

Remarcar que la clase `Node` será de uso interno (desde fuera de la EDL no se tendrá conocimiento de ella). Por tanto, la interfaz pública de `push(e)`, `pop()` y `peek()` recibe/devuelve datos (elementos de la secuencia), y no nodos. 

<div align="center">
    <img src="fig/stack_linked.png" alt="Pila con nodos enlazados" style="max-width:600px"/>
</div>

```python
class Node:
    """Nodo de una secuencia enlazada."""
    def __init__(self, data, next=None):
        self.data = data
        self.next = next


class StackLinked:
    """Pila implementada con nodos enlazados.
    
    Atributos:
        __top (Node): referencia al nodo tope de la pila.
        __size (int): número de elementos apilados.
    """
    
    def __init__(self):
        """Crea una pila vacía."""
        self.__top = None
        self.__size = 0
    
    def push(self, x):
        """Apila el elemento x en el tope de la pila."""
        self.__top = Node(x, self.__top)
        self.__size += 1
    
    def pop(self):
        """Desapila y devuelve el elemento en el tope.
        Lanza IndexError si la pila está vacía."""
        if self.__top is None:
            raise IndexError("Pila vacía")
        x = self.__top.data
        self.__top = self.__top.next
        self.__size -= 1
        return x
    
    def peek(self):
        """Devuelve el elemento en el tope sin desapilarlo.
        Lanza IndexError si la pila está vacía."""
        if self.__top is None:
            raise IndexError("Pila vacía")
        return self.__top.data
    
    def empty(self):
        """Indica si la pila está vacía."""
        return self.__top is None
    
    def size(self):
        """Devuelve el número de elementos apilados."""
        return self.__size
    
    def __str__(self):
        """Representación visual de la pila (de tope a base)."""
        result = "(StackLinked)\n"
        aux = self.__top
        while aux is not None:
            result += f"  | {aux.data} |\n"
            aux = aux.next
        result += "  +---+"
        return result
```

## 4. Implementación con array: `StackArray`

Implementamos la pila mediante una lista Python de tamaño fijo (denotado por el atributo `__MAX`) referenciada por el atributo `_arr`, donde los elementos se almacenan de la base (posición 0) al tope, este último denotado por un atributo entero `__top`. 

> Nota: el atributo `__MAX` no es estrictamente necesario (se puede calcular con `len(__arr)`), pero es una buena práctica considerarlo, ya que estamos imitando un contexto de uso de arrays puros, como sucedería con C/C++.

<div align="center">
    <img src="fig/stack_array.png" alt="Pila con array" style="max-width:550px"/>
</div>

```python
class StackArray:
    """Pila implementada con un array (lista Python de tamaño fijo).
    
    Atributos:
        __arr (list): array que almacena los elementos.
        __top (int): índice del tope (-1 si vacía).
        __MAX (int): capacidad máxima de la pila.
    """
    
    def __init__(self, max_size=100):
        """Crea una pila vacía con capacidad máxima max_size."""
        self.__MAX = max_size
        self.__arr = [None] * self.__MAX
        self.__top = -1
    
    def push(self, x):
        """Apila el elemento x. Lanza IndexError si la pila está llena."""
        if self.__top + 1 == self.__MAX:
            raise IndexError("Pila llena: no caben más elementos")
        self.__top += 1
        self.__arr[self.__top] = x
    
    def pop(self):
        """Desapila y devuelve el tope. Lanza IndexError si vacía."""
        if self.__top < 0:
            raise IndexError("Pila vacía")
        x = self.__arr[self.__top]
        self.__top -= 1
        return x
    
    def peek(self):
        """Devuelve el tope sin desapilarlo. Lanza IndexError si vacía."""
        if self.__top < 0:
            raise IndexError("Pila vacía")
        return self.__arr[self.__top]
    
    def empty(self):
        """Indica si la pila está vacía."""
        return self.__top == -1
    
    def size(self):
        """Devuelve el número de elementos apilados."""
        return self.__top + 1
    
    def __str__(self):
        result = "(StackArray)\n"
        for i in range(self.__top, -1, -1):
            result += f"  | {self.__arr[i]} |\n"
        result += "  +---+"
        return result
```

## 5. Uso y demostración

Ambas implementaciones ofrecen la **misma interfaz**, por lo que el código de uso es idéntico. Alternar entre una y otra resulta trivial:

```python
# Podemos usar cualquiera de las dos implementaciones:

pila = StackLinked() # Comentar/Descomentar
#pila = StackArray() # Comentar/Descomentar

pila.push(14)
print(f"peek(): {pila.peek()}")     # 14

pila.push(33)
print(f"peek(): {pila.peek()}")     # 33
print(f"size(): {pila.size()}")     # 2

print("\nDesapilando todos los elementos:")
while not pila.empty():
    print(f"  pop(): {pila.pop()}")  # 33, luego 14

print(f"empty(): {pila.empty()}")    # True

# Intentar desapilar una pila vacía genera una excepción
try:
    pila.pop()
except IndexError as e:
    print(f"Error: {e}")

print("\nNueva pila desde una lista:")
pila = StackLinked()
for v in [10, 22, 1, 13]:
    pila.push(v)

print(pila)
print(f"\npeek(): {pila.peek()}")
print(f"pop():  {pila.pop()}")
print(f"\nTras pop():")
print(pila)
```

## 6. Comparativa de implementaciones

| Aspecto | Nodos enlazados (`StackLinked`) | Array (`StackArray`) |
|---------|-------------------------------|---------------------|
| **Complejidad temporal** | Todas las operaciones: $\Theta(1)$ | Todas las operaciones: $\Theta(1)$ |
| **Memoria** | Variable (crece/decrece dinámicamente) + sobrecoste por referencias (despreciable) | Fija (se reserva `MAX` posiciones) |
| **Talla máxima** | Sin límite (salvo memoria del sistema) | Limitada a `MAX` |


La implementación con arrays presenta el inconveniente de la estimación del tamaño máximo. 

Se podría implementar una variante de `push()` que permita redimensionar automáticamente el array, como vimos en la [Sección 4.1 del T5.3 sobre gestión de secuencias con memoria contigua](./T5.3%20Gestión%20de%20secuencias%20con%20memoria%20contigua.ipynb), pero entonces, el coste de esa operación pasaria a ser $\Omega(1) \cap O(n)$.
