## Clase `Node` y funciones auxiliares

Para que el cuaderno sea completamente ejecutable y autónomo, definimos la clase `Node` y las funciones auxiliares para convertir listas estándar de Python a secuencias enlazadas y viceversa.


```python
class Node:
    """Nodo de una secuencia enlazada.
    
    Atributos:
        data: el dato almacenado en el nodo.
        next: referencia al siguiente nodo (None si es el último).
    """
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
    
    def __repr__(self):
        return f"Node({self.data})"

def list_to_linked(my_list):
    """Convierte una lista de Python en una secuencia de nodos enlazados."""
    seq = None
    for i in range(len(my_list) - 1, -1, -1):
        seq = Node(my_list[i], seq)
    return seq

def linked_to_str(seq):
    """Devuelve una representación en cadena de la secuencia enlazada."""
    ret = ""
    aux = seq
    while aux:
        ret += str(aux.data) + " -> "
        aux = aux.next
    ret += "NULL"
    return ret
```

## Ejercicio 1: Saturación de valores (`saturate`)

Implementa una función `saturate` que, dada una secuencia enlazada `seq` que almacena valores enteros y un entero `max_val`, limite el valor de los elementos de la secuencia a `max_val` (es decir, cualquier valor estrictamente mayor que `max_val` pasará a valer `max_val`).

*Ejemplo:*
* Secuencia original: `[1, 8, 3, 5]`
* Valor máximo: `2`
* Secuencia resultante: `[1, 2, 2, 2]`


```python
def saturate(seq, max_val):
    """Satura los valores de la secuencia enlazada seq a un valor máximo max_val.
    
    Los datos mayores que max_val se reemplazan por max_val in-place.
    """
    aux = seq
    while aux is not None:
        if aux.data > max_val:
            aux.data = max_val
        aux = aux.next

# Prueba:

lista = [1, 8, 3, 5]
max_val = 2

seq = list_to_linked(lista)
print("Secuencia original: ", linked_to_str(seq))

saturate(seq, max_val)
print("Secuencia saturada: ", linked_to_str(seq))
```

## Ejercicio 2: Inserción al final con coste lineal $\Theta(n)$

Implementa una función `append` que, dada una secuencia enlazada `seq` y un valor entero `x`, inserte `x` al final de la secuencia, con un coste temporal lineal respecto a la talla $\Theta(n)$.

```python
def append_lineal(seq, x):
    """Inserta x al final de la secuencia enlazada seq con coste lineal O(n).
    
    Retorna el primer nodo de la secuencia resultante.
    """
    nuevo_nodo = Node(x)
    
    # Caso especial: secuencia vacía
    if seq is None:
        return nuevo_nodo
    
    # Caso general: buscar el último nodo
    aux = seq
    while aux.next is not None:
        aux = aux.next
    
    # Enlazar el nuevo nodo al final
    aux.next = nuevo_nodo

    # Notar que debemos de devolver seq, ya que, si hacemos append 
    # sobre una secuencia vacía, cambia seq
    return seq

# Prueba:

print("--- Caso 1: Secuencia vacía ---")
seq_vacia = None
seq_vacia = append_lineal(seq_vacia, 10)
print("Resultado:", linked_to_str(seq_vacia))

print("\n--- Caso 2: Secuencia con elementos ---")
seq = list_to_linked([8, 2, 6])
print("Antes:    ", linked_to_str(seq))
seq = append_lineal(seq, 4)
print("Después:  ", linked_to_str(seq))
```

## Ejercicio 3: Inserción al final con coste constante $\Theta(1)$

Implementa una versión de la función `append` que inserte un elemento entero `x` al final de una secuencia de nodos enlazados con coste temporal constante respecto a la talla $\Theta(1)$. Para ello, considera un parámetro adicional `last` que será un puntero al último nodo de la secuencia.

```python
def append_constante(seq, last, x):
    """Inserta x al final de la secuencia enlazada seq con coste constante O(1).
    
    Utiliza y actualiza la referencia 'last' al último nodo.
    Retorna la tupla (seq, last) con las referencias actualizadas.
    """
    nuevo_nodo = Node(x)
    
    # Caso especial: secuencia vacía
    if last is None:
        seq = nuevo_nodo
    # Caso general: enlazar tras el último actual
    else:
        last.next = nuevo_nodo
        
    # En cualquier caso, el nuevo nodo pasa a ser el último
    last = nuevo_nodo

    # Devolvemos seq, y también last (obviamente)
    return seq, last

# Prueba:

seq = None
last = None
print("Inicialmente:", linked_to_str(seq))
seq, last = append_constante(seq, last, 10)
print("append(10):", linked_to_str(seq))
seq, last = append_constante(seq, last, 2)
print("append(2):", linked_to_str(seq))
seq, last = append_constante(seq, last, 8)
print("append(8):", linked_to_str(seq))
```

## Ejercicio 4: Inserción ordenada descendente

Implementa una función `reverse_ordered_insert` que permita insertar un elemento de forma ordenada en una secuencia de nodos enlazados, asumiendo una ordenación descendente (de mayor a menor).

```python
def reverse_ordered_insert(seq, x):
    """Inserta x manteniendo la ordenación descendente de la secuencia enlazada seq.
    
    Retorna el primer nodo de la secuencia resultante.
    """
    prev = None
    aux = seq
    
    # Buscar el primer nodo cuyo valor sea <= x
    while aux is not None and aux.data > x:
        prev = aux
        aux = aux.next
        
    # Crear e insertar el nuevo nodo
    if prev is None:  # Caso a): insertar en cabeza
        return Node(x, seq)
    else:             # Caso b): insertar entre prev y aux
        prev.next = Node(x, aux)
        return seq

# Prueba:

seq = list_to_linked([15, 10, 5, 2])
print("Secuencia ordenada original (desc): ", linked_to_str(seq))

# Inserción intermedia
seq = reverse_ordered_insert(seq, 8)
print("Insertar 8 (intermedio):            ", linked_to_str(seq))

# Inserción en cabeza
seq = reverse_ordered_insert(seq, 20)
print("Insertar 20 (cabeza):               ", linked_to_str(seq))

# Inserción al final
seq = reverse_ordered_insert(seq, 1)
print("Insertar 1 (final):                 ", linked_to_str(seq))
```

## Ejercicio 5: Eliminación de elementos en un intervalo

Implementa una función `remove_interval` que, dada una secuencia enlazada `seq` y un valor entero positivo `x`, elimine todos los elementos de la secuencia cuyo valor esté dentro del intervalo cerrado $[-x, x]$. La función debe devolver el número de elementos eliminados.

*Ejemplo:*
* Secuencia original: `[-1, 8, 3, 5, 12]`
* Valor $x$: `4` (intervalo a eliminar: `[-4, 4]`)
* Secuencia resultante: `[8, 5, 12]`
* Elementos eliminados: `2` (el `-1` y el `3`)

```python
def remove_interval(seq, x):
    """Elimina todos los nodos cuyos valores estén en el intervalo cerrado [-x, x].
    
    Retorna una tupla (seq_resultante, total_eliminados).
    """
    prev = None
    aux = seq
    count = 0
    
    while aux is not None:
        # Comprobar si el dato está en el intervalo cerrado [-x, x]
        if -x <= aux.data <= x:
            count += 1
            if aux == seq:  # Caso a): eliminar el primer nodo
                seq = seq.next
            else:           # Caso b): eliminar cualquier otro nodo (bypass)
                prev.next = aux.next
            # Avanzar solo aux; prev sigue siendo el mismo
            aux = aux.next
        else:
            # No se elimina: avanzar ambos punteros
            prev = aux
            aux = aux.next
            
    return seq, count

# Prueba:

valores = [1, 8, 3, 5, 12]
x = 4

seq = list_to_linked(valores)
print("Secuencia original:       ", linked_to_str(seq))
print(f"Intervalo a eliminar:     [{-x}, {x}]")

seq, eliminados = remove_interval(seq, x)
print("Secuencia modificada:     ", linked_to_str(seq))
print("Elementos eliminados:     ", eliminados)
```
